# Chapter 02: NumPy Arrays

NumPy (Numerical Python) is the foundational package for numerical computing in Python. It provides a high-performance multidimensional array object and tools for working with these arrays. Nearly every data science library in the Python ecosystem — pandas, scikit-learn, matplotlib, scipy — is built on top of NumPy.

In this notebook we will cover:

- Why NumPy over Python lists
- Creating arrays from Python lists
- Built-in array generation functions
- Random number generation
- Array attributes and data types
- Reshaping arrays

```python
import numpy as np
```

## Why NumPy? Python Lists vs NumPy Arrays

Python lists are flexible — they can hold mixed types and grow dynamically. However, for numerical work they have significant drawbacks:

- **Speed**: NumPy operations are implemented in C and operate on contiguous memory blocks, making them 10-100x faster than equivalent Python list operations.
- **Memory**: NumPy arrays store elements of a single type in contiguous memory, using far less memory than lists of Python objects.
- **Vectorisation**: NumPy lets you express batch operations on data without writing loops.
- **Ecosystem**: The entire scientific Python stack expects NumPy arrays.

```python
# Speed comparison: summing 1 million numbers
python_list = list(range(1_000_000))
numpy_array = np.arange(1_000_000)

%timeit sum(python_list)
%timeit np.sum(numpy_array)
```

```python
# With lists, element-wise operations require a loop or list comprehension
py_list = [1, 2, 3, 4, 5]
squared_list = [x ** 2 for x in py_list]
print("List comprehension:", squared_list)

# With NumPy, it is a single vectorised expression
arr = np.array([1, 2, 3, 4, 5])
print("NumPy vectorised: ", arr ** 2)
```

## Creating Arrays from Python Lists

The most direct way to create a NumPy array is from an existing Python list (or list of lists).

```python
# 1D array from a list
my_list = [10, 20, 30, 40, 50]
arr_1d = np.array(my_list)
arr_1d
```

```python
# 2D array from a list of lists
matrix_list = [[1, 2, 3],
               [4, 5, 6],
               [7, 8, 9]]
arr_2d = np.array(matrix_list)
arr_2d
```

```python
# 3D array
arr_3d = np.array([[[1, 2], [3, 4]],
                    [[5, 6], [7, 8]]])
print("Shape:", arr_3d.shape)
arr_3d
```

> **Note**: All elements in a NumPy array must be the same type. If you pass mixed types, NumPy will upcast to the most general type.

```python
# Mixed types are upcast
mixed = np.array([1, 2.5, 3])
print(mixed)        # [1.  2.5 3. ]
print(mixed.dtype)  # float64
```

## Built-in Array Generation Functions

NumPy provides many convenient functions to create arrays without manually specifying every element.

### `np.arange`

Similar to Python's `range()` but returns a NumPy array. Accepts `start`, `stop` (exclusive), and `step`.

```python
np.arange(0, 10)
```

```python
# With a step of 2
np.arange(0, 20, 2)
```

```python
# Float steps are allowed (unlike range)
np.arange(0, 1, 0.1)
```

### `np.zeros` and `np.ones`

Create arrays filled with zeros or ones. Pass a single integer for 1D, or a tuple for higher dimensions.

```python
np.zeros(5)
```

```python
np.zeros((3, 4))
```

```python
np.ones(4)
```

```python
np.ones((2, 3))
```

```python
# np.full lets you fill with any value
np.full((3, 3), 7)
```

### `np.linspace`

Return evenly spaced numbers over a specified interval. Unlike `arange`, you specify the **number of points** rather than the step size, and the endpoint **is included** by default.

```python
# 5 evenly spaced points between 0 and 10
np.linspace(0, 10, 5)
```

```python
# 21 points between 0 and 1 (useful for plotting)
np.linspace(0, 1, 21)
```

> **Pro Tip**: `linspace` is preferable to `arange` when you need a specific number of points in an interval, especially with floating-point values where `arange` can produce unexpected numbers of elements due to rounding.

### `np.eye`

Creates an identity matrix — a square matrix with ones on the main diagonal and zeros elsewhere. Identity matrices are fundamental in linear algebra.

```python
np.eye(4)
```

## Random Number Generation

NumPy's `random` module provides functions to generate arrays of random numbers from various distributions. This is essential for simulations, testing, and initialising machine learning models.

### `np.random.rand` — Uniform Distribution [0, 1)

Each value is drawn uniformly from the interval [0, 1).

```python
# 5 random floats
np.random.rand(5)
```

```python
# 3x3 matrix of random floats
np.random.rand(3, 3)
```

### `np.random.randn` — Standard Normal Distribution

Draws samples from a standard normal (Gaussian) distribution with mean 0 and standard deviation 1. Values near zero are more likely.

```python
np.random.randn(5)
```

```python
np.random.randn(4, 4)
```

### `np.random.randint` — Random Integers

Returns random integers from `low` (inclusive) to `high` (exclusive).

```python
# Single random integer between 1 and 100
np.random.randint(1, 100)
```

```python
# Array of 10 random integers between 1 and 100
np.random.randint(1, 100, 10)
```

```python
# 3x4 matrix of random integers
np.random.randint(0, 50, size=(3, 4))
```

### `np.random.seed` — Reproducibility

Setting a seed ensures the same sequence of random numbers is generated every time. This is critical for reproducible experiments and debugging.

```python
np.random.seed(42)
print("First call: ", np.random.rand(4))

np.random.seed(42)
print("Second call:", np.random.rand(4))  # Identical output
```

> **Pro Tip**: In modern NumPy (1.17+), the recommended approach is to use a `Generator` object via `np.random.default_rng()` instead of the legacy `np.random.seed()` API. The new API is more flexible and avoids global state issues.

```python
# Modern approach with Generator
rng = np.random.default_rng(seed=42)
print(rng.random(5))       # uniform [0, 1)
print(rng.integers(0, 10, size=5))  # random ints
```

## Array Attributes

Every NumPy array carries metadata about its structure. Understanding these attributes is essential for debugging shape mismatches and understanding memory usage.

```python
arr = np.random.randint(0, 100, size=(3, 5))
arr
```

```python
# shape: tuple of dimension sizes
print("shape:   ", arr.shape)     # (3, 5)

# ndim: number of dimensions (axes)
print("ndim:    ", arr.ndim)      # 2

# size: total number of elements
print("size:    ", arr.size)      # 15

# dtype: data type of elements
print("dtype:   ", arr.dtype)     # int64

# itemsize: bytes per element
print("itemsize:", arr.itemsize)  # 8 (bytes for int64)

# nbytes: total memory in bytes (size * itemsize)
print("nbytes:  ", arr.nbytes)    # 120
```

## Reshaping Arrays

Reshaping changes the layout of an array without changing its data. The total number of elements must remain the same.

```python
arr = np.arange(12)
print("Original:", arr)
print("Shape:   ", arr.shape)
```

```python
# Reshape to 3 rows x 4 columns
arr.reshape(3, 4)
```

```python
# Reshape to 4 rows x 3 columns
arr.reshape(4, 3)
```

```python
# Use -1 to let NumPy infer one dimension
arr.reshape(2, -1)  # 2 rows, NumPy figures out 6 columns
```

```python
# Reshape to a column vector
arr.reshape(-1, 1)
```

```python
# Flatten back to 1D
matrix = arr.reshape(3, 4)
print("Flattened with .ravel():  ", matrix.ravel())   # returns a view when possible
print("Flattened with .flatten():", matrix.flatten())  # always returns a copy
```

> **Note**: `reshape` will raise a `ValueError` if the new shape is not compatible with the total number of elements.

```python
# This will fail: 12 elements cannot fill a 3x5 shape
try:
    arr.reshape(3, 5)
except ValueError as e:
    print(f"ValueError: {e}")
```

## Data Types

NumPy supports a rich set of numeric data types, giving you fine-grained control over precision and memory usage.

| Type | Description |
|------|------------|
| `int8`, `int16`, `int32`, `int64` | Signed integers of various sizes |
| `uint8`, `uint16`, `uint32`, `uint64` | Unsigned integers |
| `float16`, `float32`, `float64` | Floating-point numbers |
| `complex64`, `complex128` | Complex numbers |
| `bool` | Boolean (True/False) |

```python
# Specify dtype at creation
arr_float32 = np.array([1, 2, 3], dtype=np.float32)
print(arr_float32, "- dtype:", arr_float32.dtype, "- itemsize:", arr_float32.itemsize, "bytes")

arr_int8 = np.array([1, 2, 3], dtype=np.int8)
print(arr_int8, "- dtype:", arr_int8.dtype, "- itemsize:", arr_int8.itemsize, "byte")
```

```python
# Convert dtype with .astype()
arr_int = np.array([1, 2, 3, 4])
arr_float = arr_int.astype(np.float64)
print(arr_float, "- dtype:", arr_float.dtype)
```

```python
# Boolean arrays
arr_bool = np.array([True, False, True, True])
print(arr_bool.dtype)  # bool
print(arr_bool.sum())  # True counts as 1 -> sum = 3
```

> **Pro Tip**: In data science, `float64` is the default and usually fine. When working with very large datasets or image data (pixel values 0-255), using smaller types like `float32` or `uint8` can cut memory usage significantly.
