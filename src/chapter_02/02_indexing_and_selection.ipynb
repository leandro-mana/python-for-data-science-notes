```python
import numpy as np
```

## 1D Indexing and Slicing

NumPy arrays support the same indexing and slicing syntax as Python lists: `arr[start:stop:step]`. Indexing is zero-based, and negative indices count from the end.

```python
arr = np.arange(10, 21)  # [10, 11, 12, ..., 20]
arr
```

```python
# Single element access
print("First element: ", arr[0])
print("Last element:  ", arr[-1])
print("Fifth element: ", arr[4])
```

```python
# Slicing: start:stop (stop is exclusive)
print("arr[2:6] =", arr[2:6])   # elements at index 2,3,4,5
print("arr[:4]  =", arr[:4])    # first 4 elements
print("arr[7:]  =", arr[7:])    # from index 7 to the end
```

```python
# Slicing with step
print("Every 2nd element: ", arr[::2])
print("Reversed:          ", arr[::-1])
print("Every 3rd from idx 1:", arr[1::3])
```

## 2D Indexing

For 2D arrays (matrices), you can index using `arr[row, col]` or `arr[row][col]`. The comma notation is preferred for clarity and performance.

![NumPy indexing reference](data/numpy_indexing.png)

```python
arr_2d = np.array([[5, 10, 15],
                   [20, 25, 30],
                   [35, 40, 45]])
arr_2d
```

```python
# Select a single row
print("Row 0:", arr_2d[0])
print("Row 2:", arr_2d[2])
```

```python
# Select a single element
print("Row 1, Col 2:", arr_2d[1, 2])  # 30
print("Row 0, Col 0:", arr_2d[0, 0])  # 5
```

```python
# Select a single column
print("Column 1:", arr_2d[:, 1])  # [10, 25, 40]
```

```python
# Submatrix slicing: top-right 2x2 block
arr_2d[:2, 1:]
```

```python
# Bottom-left 2x2 block
arr_2d[1:, :2]
```

## Fancy Indexing

Fancy indexing lets you pass an array of indices to select specific elements or rows in any order. Unlike slicing, fancy indexing always returns a **copy**, not a view.

```python
arr = np.arange(0, 100, 10)  # [0, 10, 20, ..., 90]
print("Array:", arr)

# Select elements at indices 2, 5, and 7
indices = [2, 5, 7]
print("Fancy indexed:", arr[indices])
```

```python
# Fancy indexing with 2D arrays — select specific rows
matrix = np.arange(1, 17).reshape(4, 4)
print("Matrix:")
print(matrix)
print()

# Select rows 0, 2, 3 (in that order)
print("Rows [0, 2, 3]:")
print(matrix[[0, 2, 3]])
```

```python
# Select rows in reverse order
print("Rows reversed:")
print(matrix[[3, 2, 1, 0]])
```

```python
# Select specific elements: (row0,col1), (row1,col3), (row2,col0)
print("Specific elements:", matrix[[0, 1, 2], [1, 3, 0]])
```

## Boolean / Conditional Selection

One of NumPy's most powerful features is the ability to filter arrays using boolean conditions. A comparison on an array produces a boolean array, which can then be used as an index to select matching elements.

This concept translates directly to pandas, so make sure you understand it well.

```python
arr = np.arange(1, 11)
print("Array:", arr)
```

```python
# Comparison produces a boolean array
print("arr > 5:", arr > 5)
```

```python
# Use the boolean array to filter
arr[arr > 5]
```

```python
# Store the boolean mask and reuse it
mask = arr % 2 == 0  # even numbers
print("Even mask:", mask)
print("Even values:", arr[mask])
```

```python
# Combine conditions with & (and), | (or), ~ (not)
# Important: use parentheses around each condition!
print("Between 3 and 8:", arr[(arr >= 3) & (arr <= 8)])
print("Less than 3 or greater than 8:", arr[(arr < 3) | (arr > 8)])
print("Not equal to 5:", arr[arr != 5])
```

> **Pro Tip**: Always use `&`, `|`, `~` (bitwise operators) for element-wise boolean operations on arrays, NOT `and`, `or`, `not` (Python logical operators). The logical operators only work with single booleans and will raise an error on arrays.

## Setting Values with Indexing

You can assign new values using any of the indexing methods above.

```python
arr = np.arange(0, 11)
print("Before:", arr)

# Set a single value
arr[0] = 99
print("After arr[0]=99:", arr)
```

```python
# Broadcasting a single value to a slice
arr = np.arange(0, 11)
arr[0:5] = 100
print("After arr[0:5]=100:", arr)
```

```python
# Conditional assignment
arr = np.arange(0, 11)
arr[arr < 5] = 0
print("After zeroing values < 5:", arr)
```

## Views vs Copies

This is a critical concept in NumPy. Slicing an array returns a **view** — a window into the original data, not a separate copy. Modifying the view modifies the original array.

This design avoids unnecessary memory copies, but it can catch you off guard.

```python
original = np.arange(10)
print("Original:", original)

# Create a slice (this is a VIEW)
slice_view = original[3:7]
print("Slice:   ", slice_view)
```

```python
# Modify the view
slice_view[:] = 999

# The original is affected!
print("Slice:   ", slice_view)
print("Original:", original)
```

```python
# To avoid this, explicitly make a copy
original = np.arange(10)
safe_copy = original[3:7].copy()

safe_copy[:] = 999
print("Copy:    ", safe_copy)
print("Original:", original)  # unchanged
```

**When do you get a view vs a copy?**

| Operation | Returns |
|-----------|--------|
| Basic slicing (`arr[2:5]`) | View |
| Fancy indexing (`arr[[1,3,5]]`) | Copy |
| Boolean indexing (`arr[arr > 0]`) | Copy |
| `.copy()` | Copy |
| `.reshape()` | View (when possible) |

## `np.where` — Conditional Logic on Arrays

`np.where(condition, x, y)` returns an array where elements are taken from `x` where the condition is True, and from `y` where it is False. Think of it as a vectorised if/else.

```python
arr = np.array([1, -2, 3, -4, 5, -6])

# Replace negatives with 0, keep positives
result = np.where(arr > 0, arr, 0)
print("Clipped to positive:", result)
```

```python
# Label as 'positive' or 'negative'
labels = np.where(arr > 0, 'positive', 'negative')
print(labels)
```

```python
# np.where with only condition returns INDICES where True
indices = np.where(arr > 0)
print("Indices where positive:", indices[0])
print("Values at those indices:", arr[indices])
```

```python
# Practical example: cap exam scores at 100
scores = np.array([85, 92, 105, 78, 110, 95])
capped = np.where(scores > 100, 100, scores)
print("Original:", scores)
print("Capped:  ", capped)
```

## Working with NaN Values

`np.nan` (Not a Number) is a special float value used to represent missing or undefined data. It is common in real-world datasets.

```python
data = np.array([1.0, 2.0, np.nan, 4.0, np.nan, 6.0])
print("Data:", data)
```

```python
# NaN propagates through calculations
print("Sum (with nan):", np.sum(data))     # nan
print("Mean (with nan):", np.mean(data))   # nan
```

```python
# Use nan-safe functions to ignore NaN values
print("nansum: ", np.nansum(data))
print("nanmean:", np.nanmean(data))
print("nanstd: ", np.nanstd(data))
```

```python
# Detect NaN values
print("Is NaN?:", np.isnan(data))

# Count NaN values
print("Number of NaNs:", np.isnan(data).sum())
```

```python
# Filter out NaN values
clean_data = data[~np.isnan(data)]
print("Without NaNs:", clean_data)
```

```python
# Replace NaN with a value using np.where or np.nan_to_num
filled = np.where(np.isnan(data), 0, data)
print("NaN replaced with 0:  ", filled)

filled2 = np.nan_to_num(data, nan=-1.0)
print("NaN replaced with -1: ", filled2)
```

> **Pro Tip**: NaN comparisons are tricky — `np.nan == np.nan` is `False`. Always use `np.isnan()` to test for NaN, never `==`.

```python
# NaN comparison pitfall
print("np.nan == np.nan:", np.nan == np.nan)  # False!
print("np.isnan(np.nan):", np.isnan(np.nan))  # True
```

## Summary

- **1D slicing** works like Python lists: `arr[start:stop:step]`.
- **2D indexing** uses `arr[row, col]` notation; slicing works on both axes.
- **Fancy indexing** uses integer arrays to select elements in arbitrary order (returns a copy).
- **Boolean selection** uses comparison operators to create masks and filter arrays.
- **Views vs copies**: slicing creates views (shared memory); fancy/boolean indexing creates copies.
- **`np.where`** provides vectorised if/else logic.
- **NaN handling** requires special functions (`np.isnan`, `np.nanmean`, etc.).
