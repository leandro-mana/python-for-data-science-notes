# Chapter 03: Text Methods

Real-world text data is messy — inconsistent casing, extra whitespace, embedded patterns that need extraction. Pandas exposes Python string methods through the `.str` accessor, allowing vectorised string operations across an entire Series.

In this notebook we will cover:

- The `.str` accessor
- Case conversion: `.str.lower()`, `.str.upper()`, `.str.title()`, `.str.strip()`
- Searching: `.str.contains()`, `.str.startswith()`, `.str.endswith()`
- Manipulation: `.str.replace()`, `.str.split()`, `.str.get()`
- Info: `.str.len()`, `.str.count()`
- Regex extraction: `.str.extract()`
- Joining: `.str.cat()`

```python
import pandas as pd
import numpy as np
```

```python
tips = pd.read_csv('data/tips.csv')
tips.head()
```

## The `.str` Accessor

The `.str` accessor is available on any Series with `object` or `string` dtype. It provides vectorised versions of Python's built-in string methods.

```python
names = tips['Payer Name']
names.head()
```

## Case Conversion

```python
names.str.lower().head()
```

```python
names.str.upper().head()
```

```python
# title() capitalises the first letter of each word
messy = pd.Series(['  alice SMITH  ', 'BOB jones', 'charlie BROWN'])
messy.str.title()
```

## Stripping Whitespace

```python
# strip removes leading and trailing whitespace
messy.str.strip()
```

```python
# lstrip / rstrip for one side only
print(messy.str.lstrip().values)
print(messy.str.rstrip().values)
```

```python
# Chaining: clean up messy names
messy.str.strip().str.title()
```

## Searching: `.str.contains()`, `.str.startswith()`, `.str.endswith()`

```python
# Find payers whose name contains 'John'
mask = names.str.contains('John', na=False)
tips.loc[mask, ['Payer Name', 'total_bill']].head()
```

```python
# Case-insensitive search
names.str.contains('john', case=False, na=False).sum()
```

```python
# startswith and endswith
print(f"Names starting with 'A': {names.str.startswith('A', na=False).sum()}")
print(f"Names ending with 'son': {names.str.endswith('son', na=False).sum()}")
```

```python
# Use contains with regex (default is regex=True)
# Find names with exactly two words
names.str.contains(r'^\S+\s\S+$', na=False).sum()
```

## `.str.replace()`

```python
# Replace a literal string
emails = pd.Series(['alice@gmail.com', 'bob@yahoo.com', 'charlie@gmail.com'])
emails.str.replace('gmail.com', 'company.com', regex=False)
```

```python
# Replace with regex: remove digits from names
dirty_names = pd.Series(['Alice123', 'Bob456', 'Charlie789'])
dirty_names.str.replace(r'\d+', '', regex=True)
```

## `.str.split()` and `.str.get()`

```python
# Split payer names into first and last
names.str.split(' ').head()
```

```python
# Get the first name (index 0)
names.str.split(' ').str.get(0).head()
```

```python
# Get the last name (index -1)
names.str.split(' ').str.get(-1).head()
```

```python
# expand=True splits into separate columns (returns a DataFrame)
names.str.split(' ', expand=True).head()
```

## `.str.len()` and `.str.count()`

```python
# Length of each name
names.str.len().head(10)
```

```python
# Count occurrences of a character
names.str.count('a').head(10)
```

```python
# Count vowels using regex
names.str.count(r'[aeiouAEIOU]').head(10)
```

## `.str.extract()` — Regex Group Extraction

`.str.extract()` pulls out the first match of a regex group into new columns.

```python
# Extract first and last name into separate columns
names.str.extract(r'(\w+)\s+(\w+)').head()
```

```python
# Named groups become column names
names.str.extract(r'(?P<first>\w+)\s+(?P<last>\w+)').head()
```

```python
# Extract domain from email addresses
emails = pd.Series(['alice@gmail.com', 'bob@yahoo.co.uk', 'charlie@company.org'])
emails.str.extract(r'@(.+)')
```

## `.str.cat()` — Concatenating Strings

`.str.cat()` joins all strings in a Series, or combines two Series element-wise.

```python
# Join all names into a single string
fruits = pd.Series(['apple', 'banana', 'cherry'])
fruits.str.cat(sep=', ')
```

```python
# Element-wise concatenation of two Series
first = pd.Series(['Alice', 'Bob', 'Charlie'])
last = pd.Series(['Smith', 'Jones', 'Brown'])

first.str.cat(last, sep=' ')
```

```python
# Handle NaN in concatenation
s = pd.Series(['hello', np.nan, 'world'])
s.str.cat(sep=' ', na_rep='???')
```

## Practical Example: Cleaning the MPG Dataset Names

```python
mpg = pd.read_csv('data/mpg.csv')

# Extract the manufacturer (first word) from the car name
mpg['manufacturer'] = mpg['name'].str.split().str.get(0).str.title()

# Top 10 manufacturers by count
mpg['manufacturer'].value_counts().head(10)
```

```python
# Find all cars with 'toyota' in the name
mpg[mpg['name'].str.contains('toyota', case=False)][['name', 'mpg', 'model_year']].head(10)
```

## Summary

In this notebook we covered text methods in pandas:

- The `.str` accessor provides vectorised string operations on Series.
- Case: `.str.lower()`, `.str.upper()`, `.str.title()`; whitespace: `.str.strip()`.
- Searching: `.str.contains()`, `.str.startswith()`, `.str.endswith()` — all support regex.
- `.str.replace()` substitutes patterns; `.str.split()` tokenises strings.
- `.str.get()` indexes into split results; `expand=True` returns a DataFrame.
- `.str.extract()` pulls regex capture groups into new columns.
- `.str.cat()` joins strings within a Series or across two Series.
