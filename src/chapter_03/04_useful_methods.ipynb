# Chapter 03: Useful Methods

Pandas provides a rich set of built-in methods that cover the vast majority of data manipulation tasks. Mastering these methods means fewer custom loops and faster, more readable code.

In this notebook we will cover:

- `apply()` on columns and rows
- `map()` for element-wise transformations
- `sort_values()` and `sort_index()`
- `value_counts()`, `nunique()`, `unique()`
- `.duplicated()` and `.drop_duplicates()`
- `.replace()`, `.rename()`
- `.sample()`, `.nlargest()`, `.nsmallest()`
- `.clip()`, `.abs()`

```python
import pandas as pd
import numpy as np
```

```python
tips = pd.read_csv('data/tips.csv')
tips.head()
```

## `apply()` — Apply a Function Along an Axis

`apply()` lets you run any function across columns (`axis=0`, default) or rows (`axis=1`). It is the Swiss-army knife for custom transformations.

```python
# Apply a function to a single column (Series)
tips['total_bill'].apply(lambda x: round(x, 0)).head()
```

```python
# Named function example
def tip_quality(tip):
    if tip >= 5:
        return 'generous'
    elif tip >= 2:
        return 'normal'
    else:
        return 'low'

tips['tip_quality'] = tips['tip'].apply(tip_quality)
tips[['tip', 'tip_quality']].head(10)
```

```python
# Apply across rows (axis=1): each row is passed as a Series
def describe_meal(row):
    return f"{row['time']} on {row['day']} for {row['size']} people"

tips['meal_desc'] = tips.apply(describe_meal, axis=1)
tips['meal_desc'].head()
```

> **Performance note**: `apply()` is flexible but iterates row-by-row in Python. Vectorised operations (e.g., `tips['tip'] / tips['total_bill']`) are always faster.

## `map()` — Element-wise Transformation on a Series

`map()` works on a Series and can take a function, dictionary, or another Series for value substitution.

```python
# Map with a dictionary to recode values
day_map = {'Sun': 'Weekend', 'Sat': 'Weekend', 'Thur': 'Weekday', 'Fri': 'Weekday'}
tips['day_type'] = tips['day'].map(day_map)
tips[['day', 'day_type']].head(10)
```

```python
# Map with a function (same as apply on a Series)
tips['sex'].map({'Female': 'F', 'Male': 'M'}).head()
```

## `sort_values()` and `sort_index()`

```python
# Sort by a single column (ascending by default)
tips.sort_values('total_bill').head()
```

```python
# Sort descending
tips.sort_values('total_bill', ascending=False).head()
```

```python
# Sort by multiple columns
tips.sort_values(['day', 'total_bill'], ascending=[True, False]).head(10)
```

```python
# sort_index sorts by the index labels
shuffled = tips.sample(frac=1, random_state=42)  # shuffle all rows
shuffled.sort_index().head()  # restore original order
```

## `value_counts()`, `nunique()`, `unique()`

```python
# value_counts: frequency of each unique value
tips['day'].value_counts()
```

```python
# Normalised to proportions
tips['day'].value_counts(normalize=True)
```

```python
# unique: array of distinct values
tips['day'].unique()
```

```python
# nunique: count of distinct values
tips['day'].nunique()
```

```python
# nunique across all columns
tips.nunique()
```

## `.duplicated()` and `.drop_duplicates()`

```python
# Create a small example with duplicates
sample = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob'],
    'score': [90, 85, 90, 78, 85]
})
sample
```

```python
# duplicated returns a boolean Series (True for duplicate rows)
sample.duplicated()
```

```python
# Drop duplicate rows
sample.drop_duplicates()
```

```python
# Check duplicates on a subset of columns
sample.drop_duplicates(subset=['name'])
```

```python
# Keep the last occurrence instead of the first
sample.drop_duplicates(subset=['name'], keep='last')
```

## `.replace()` and `.rename()`

```python
# Replace specific values
tips['sex'].replace({'Female': 'F', 'Male': 'M'}).head()
```

```python
# Rename columns
tips.rename(columns={'total_bill': 'bill', 'tip': 'gratuity'}).head(3)
```

```python
# Rename with a function (e.g., make all column names lowercase)
tips.rename(columns=str.lower).columns
```

## `.sample()`, `.nlargest()`, `.nsmallest()`

```python
# Random sample of 5 rows
tips.sample(n=5, random_state=42)
```

```python
# Random fraction (10% of the data)
tips.sample(frac=0.1, random_state=0).shape
```

```python
# Top 5 bills by total_bill
tips.nlargest(5, 'total_bill')[['total_bill', 'tip', 'Payer Name']]
```

```python
# Bottom 5 tips
tips.nsmallest(5, 'tip')[['total_bill', 'tip', 'Payer Name']]
```

## `.clip()` and `.abs()`

```python
# clip constrains values to a range
tips['total_bill'].clip(lower=10, upper=40).describe()
```

```python
# abs returns absolute values (useful for error or residual calculations)
errors = pd.Series([-3.2, 1.5, -0.8, 4.1, -2.3])
errors.abs()
```

## Summary

In this notebook we covered essential pandas methods:

- `apply()` runs custom functions across columns or rows; prefer vectorised operations when possible.
- `map()` transforms individual elements using a function or dictionary.
- `sort_values()` and `sort_index()` sort by column values or index labels.
- `value_counts()`, `unique()`, and `nunique()` summarise categorical data.
- `duplicated()` and `drop_duplicates()` detect and remove duplicate rows.
- `replace()` substitutes values; `rename()` changes column or index labels.
- `sample()` draws random rows; `nlargest()` / `nsmallest()` return top/bottom entries.
- `clip()` bounds values to a range; `abs()` returns absolute values.
