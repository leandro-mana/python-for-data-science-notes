# Chapter 03: DataFrames

The **DataFrame** is the primary two-dimensional data structure in pandas. It represents tabular data with rows and columns — much like a spreadsheet or SQL table. Each column is a Series, and all columns share a common index.

In this notebook we will cover:

- Creating DataFrames from dicts, lists, and NumPy arrays
- Column selection (single and multiple)
- Adding and removing columns
- Row selection with `.loc` and `.iloc`
- Setting and resetting the index
- Basic inspection: `info()`, `describe()`, `dtypes`, `shape`

```python
import pandas as pd
import numpy as np
```

## Creating DataFrames from a Dictionary

The most common way to create a DataFrame is from a dictionary where keys become column names and values are lists (or arrays) of equal length.

```python
data = {
    'city': ['Sydney', 'Melbourne', 'Brisbane', 'Perth', 'Adelaide'],
    'state': ['NSW', 'VIC', 'QLD', 'WA', 'SA'],
    'population': [5_312_000, 5_078_000, 2_560_000, 2_192_000, 1_388_000],
    'area_km2': [12_368, 9_993, 15_826, 6_418, 3_258]
}

df = pd.DataFrame(data)
df
```

## Creating DataFrames from a List of Lists

Each inner list becomes a row. You supply column names separately.

```python
rows = [
    ['Alice', 28, 72_000],
    ['Bob', 34, 88_000],
    ['Charlie', 45, 120_000],
    ['Diana', 31, 95_000]
]

employees = pd.DataFrame(rows, columns=['name', 'age', 'salary'])
employees
```

## Creating DataFrames from NumPy Arrays

```python
rng = np.random.default_rng(42)
arr = rng.integers(50, 100, size=(4, 3))

scores = pd.DataFrame(arr, columns=['maths', 'science', 'english'],
                       index=['student_1', 'student_2', 'student_3', 'student_4'])
scores
```

## Loading a DataFrame from CSV

In practice, most DataFrames come from external files. Let us load the tips dataset for the rest of this notebook.

```python
tips = pd.read_csv('data/tips.csv')
tips.head()
```

Always start by understanding the shape and structure of your data.

```python
# Shape: (rows, columns)
tips.shape
```

```python
# Concise summary: non-null counts, dtypes, memory usage
tips.info()
```

```python
# Column listing
tips.columns
```

## Column Selection

### Single Column

Selecting a single column returns a **Series**.

```python
# Bracket notation (always works)
tips['total_bill'].head()
```

```python
# Dot notation (only works if column name is a valid Python identifier
# and does not clash with a DataFrame method)
tips.total_bill.head()
```

### Multiple Columns

Passing a list of column names returns a **DataFrame**.

```python
tips[['total_bill', 'tip', 'size']].head()
```

## Adding New Columns

You can create new columns through simple assignment. The new column is computed from existing ones.

```python
# Tip percentage
tips['tip_pct'] = (tips['tip'] / tips['total_bill']) * 100
tips[['total_bill', 'tip', 'tip_pct']].head()
```

```python
# Bill per person
tips['bill_per_person'] = tips['total_bill'] / tips['size']
tips[['total_bill', 'size', 'bill_per_person']].head()
```

## Removing Columns

Use `.drop()` with `axis=1` (or `columns=`) to remove columns. This returns a new DataFrame by default.

```python
# Drop a single column
tips_clean = tips.drop(columns=['bill_per_person'])
tips_clean.columns
```

```python
# Drop multiple columns
tips_slim = tips.drop(columns=['tip_pct', 'bill_per_person', 'CC Number', 'Payment ID'])
tips_slim.head()
```

## Row Selection with `.loc` and `.iloc`

Just like with Series:

- **`.loc[row_label, col_label]`** — label-based (inclusive of stop)
- **`.iloc[row_pos, col_pos]`** — integer position-based (exclusive of stop)

```python
# .iloc: get the first row as a Series
tips.iloc[0]
```

```python
# .iloc: slice rows and select specific columns by position
tips.iloc[0:5, 0:3]
```

```python
# .loc: select rows by label (here our index is integers, so labels match positions)
tips.loc[0:4, ['total_bill', 'tip', 'sex']]
```

```python
# .loc: select a single cell
tips.loc[0, 'Payer Name']
```

## Removing Rows

Use `.drop()` with the index labels. By default `axis=0` (rows).

```python
# Drop rows by index label
tips_subset = tips.drop(index=[0, 1, 2])
tips_subset.head()
```

## Setting and Resetting the Index

By default, DataFrames have a numeric `RangeIndex`. You can promote a column to become the index with `.set_index()`, and reverse it with `.reset_index()`.

```python
# Set 'Payer Name' as the index
tips_indexed = tips.set_index('Payer Name')
tips_indexed.head()
```

```python
# Now .loc uses the Payer Name as the label
tips_indexed.loc['Douglas Tucker']
```

```python
# Reset the index back to default
tips_reset = tips_indexed.reset_index()
tips_reset.head()
```

> **Note**: Both `.set_index()` and `.reset_index()` return new DataFrames by default. Pass `inplace=True` to modify in place, though the explicit reassignment pattern (`df = df.set_index(...)`) is generally preferred.

## Summary

In this notebook we covered the pandas **DataFrame**:

- DataFrames can be created from dictionaries, lists of lists, NumPy arrays, or loaded from files.
- Always inspect new data with `shape`, `dtypes`, `info()`, and `describe()`.
- Select columns with `df['col']` (Series) or `df[['col1', 'col2']]` (DataFrame).
- Add columns with `df['new'] = ...` and remove them with `df.drop(columns=...)`.
- Use `.loc` for label-based and `.iloc` for position-based row/column selection.
- `.set_index()` promotes a column to the index; `.reset_index()` reverses it.
