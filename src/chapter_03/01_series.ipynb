# Chapter 03: Pandas Series

A **Series** is the fundamental one-dimensional data structure in pandas. It is essentially a labelled array — it holds a sequence of values together with an associated array of labels called the **index**. Understanding Series is critical because every column in a pandas DataFrame is a Series.

In this notebook we will cover:

- Creating Series from lists, dicts, and NumPy arrays
- The `name` attribute
- Indexing and slicing (label-based and position-based)
- `.loc` and `.iloc` on Series
- Operations between Series (alignment on index, NaN handling)

```python
import pandas as pd
import numpy as np
```

## Creating a Series from a Python List

The simplest way to create a Series is from a Python list. By default pandas assigns a `RangeIndex` (0, 1, 2, ...) as the index.

```python
temperatures = pd.Series([22.1, 25.4, 19.8, 30.2, 27.5])
temperatures
```

```python
# You can provide a custom index with the index parameter
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
temperatures = pd.Series([22.1, 25.4, 19.8, 30.2, 27.5], index=days)
temperatures
```

## Creating a Series from a Dictionary

When you pass a dictionary, the keys become the index labels and the values become the Series values. This is a natural mapping since both structures associate labels with values.

```python
population = pd.Series({
    'Australia': 26_000_000,
    'Brazil': 214_000_000,
    'Canada': 39_000_000,
    'Germany': 84_000_000,
    'Japan': 125_000_000
})
population
```

## Creating a Series from a NumPy Array

Since pandas is built on top of NumPy, creating a Series from an ndarray is straightforward. The underlying data is stored as a NumPy array.

```python
arr = np.random.default_rng(42).standard_normal(5)
s = pd.Series(arr, index=['a', 'b', 'c', 'd', 'e'])
s
```

```python
# The values attribute returns the underlying NumPy array
print(type(s.values))
s.values
```

```python
# The index attribute returns the index object
s.index
```

## Named Series

A Series can carry a `name` attribute. This becomes especially important when a Series is part of a DataFrame — the `name` is used as the column header.

```python
revenue = pd.Series(
    [150_000, 230_000, 180_000, 310_000],
    index=['Q1', 'Q2', 'Q3', 'Q4'],
    name='quarterly_revenue'
)
revenue
```

```python
print(revenue.name)
```

```python
# You can rename a Series after creation
revenue_renamed = revenue.rename('income')
print(revenue_renamed.name)
```

## Label-based Indexing

When a Series has a custom index, you can access elements using the label directly.

```python
population
```

```python
# Access a single element by label
population['Japan']
```

```python
# Access multiple elements by passing a list of labels
population[['Australia', 'Canada', 'Germany']]
```

```python
# Label-based slicing is INCLUSIVE of the stop label
population['Brazil':'Japan']
```

## Position-based Indexing

Even with custom labels, you can still access elements by their integer position.

```python
# Integer-based slicing (stop is EXCLUSIVE, like standard Python)
population[1:4]
```

## `.loc` and `.iloc`

To avoid ambiguity — especially when your index contains integers — pandas provides two explicit accessors:

- **`.loc`** — **label-based** indexing (inclusive of stop)
- **`.iloc`** — **integer position-based** indexing (exclusive of stop)

> **Best practice**: Always prefer `.loc` or `.iloc` over bare bracket indexing `[]` to make your intent explicit.

```python
# .loc: label-based
population.loc['Brazil']
```

```python
# .loc slice is INCLUSIVE of stop
population.loc['Australia':'Germany']
```

```python
# .iloc: integer position-based
population.iloc[0]
```

```python
# .iloc slice is EXCLUSIVE of stop (standard Python behaviour)
population.iloc[1:4]
```

```python
# Why this matters: integer index can be confusing without .loc / .iloc
confusing = pd.Series([100, 200, 300], index=[3, 1, 2])
print("confusing.loc[1]:  ", confusing.loc[1])   # label 1 -> 200
print("confusing.iloc[1]: ", confusing.iloc[1])   # position 1 -> 200 (coincidence here)
print("confusing.loc[3]:  ", confusing.loc[3])   # label 3 -> 100
print("confusing.iloc[0]: ", confusing.iloc[0])   # position 0 -> 100
```

## Boolean Indexing on Series

You can filter a Series by passing a boolean condition.

```python
# Countries with population over 50 million
population[population > 50_000_000]
```

```python
# Check membership
'Japan' in population
```

## Operations Between Series

One of the most powerful features of pandas is **automatic index alignment**. When you perform arithmetic between two Series, pandas aligns them on their index labels. Labels that exist in one Series but not the other produce `NaN`.

```python
gdp = pd.Series({
    'Australia': 1.7e12,
    'Brazil': 1.9e12,
    'Canada': 2.1e12,
    'France': 2.9e12,    # Not in population
    'Germany': 4.3e12,
})
gdp
```

```python
# GDP per capita: pandas aligns on index automatically
gdp_per_capita = gdp / population
gdp_per_capita
```

Notice that:
- **France** appears only in `gdp`, so the result is `NaN`
- **Japan** appears only in `population`, so the result is also `NaN`
- Countries present in both Series get a computed value

```python
# You can handle NaN using .add() with fill_value to substitute a default
s1 = pd.Series({'a': 1, 'b': 2, 'c': 3})
s2 = pd.Series({'b': 10, 'c': 20, 'd': 30})

print("Default addition (NaN for missing):")
print(s1 + s2)
print()
print("Using .add() with fill_value=0:")
print(s1.add(s2, fill_value=0))
```

## Scalar Operations (Broadcasting)

Arithmetic with a scalar is applied element-wise, just like NumPy.

```python
# Convert population to millions
population / 1_000_000
```

```python
# Common aggregations
print(f"Mean population: {population.mean():,.0f}")
print(f"Max population:  {population.max():,.0f}")
print(f"Sum:             {population.sum():,.0f}")
```

## Summary

In this notebook we covered the pandas **Series**:

- A Series is a one-dimensional labelled array, created from lists, dicts, or NumPy arrays.
- The `name` attribute identifies the Series and becomes the column header in a DataFrame.
- Use `.loc` for label-based access and `.iloc` for position-based access — always prefer these over bare `[]` to be explicit.
- Arithmetic between Series automatically aligns on index labels, introducing `NaN` where labels do not match.
- The `.add()`, `.sub()`, `.mul()`, `.div()` methods accept a `fill_value` parameter to handle missing labels gracefully.
