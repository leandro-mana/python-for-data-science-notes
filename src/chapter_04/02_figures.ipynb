# Figures and the Object-Oriented Interface

Matplotlib has two interfaces:

1. **Functional (pyplot)** — `plt.plot()`, `plt.xlabel()`, etc. Quick but limited.
2. **Object-Oriented (OO)** — Create `Figure` and `Axes` objects explicitly and call methods on them.

The OO approach is more verbose but gives full control over every element of the plot. It is the recommended approach for anything beyond a quick one-off plot.

````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
````

## Core Concepts: Figure vs Axes

- **Figure** — The entire window or image. Think of it as the canvas.
- **Axes** — A single plot area within the figure. A figure can contain one or many axes.

```
┌──────────────── Figure ────────────────┐
│                                        │
│   ┌──────── Axes ────────┐             │
│   │  title               │             │
│   │  ┌───────────────┐   │             │
│   │  │               │   │             │
│   │  │   plot area   │   │             │
│   │  │               │   │             │
│   │  └───────────────┘   │             │
│   │  xlabel               │             │
│   └──────────────────────┘             │
│                                        │
└────────────────────────────────────────┘
```

**Key distinction:** "Axes" is not the plural of "axis." An `Axes` object is a single plot containing two (or three) `Axis` objects (x-axis, y-axis, and optionally z-axis).

## Method 1: `plt.subplots()` (Recommended)

The most common way to create a figure and axes together. Returns both objects in a single call.

````python
# Create a figure with a single axes
fig, ax = plt.subplots()

x = np.linspace(0, 2 * np.pi, 200)
ax.plot(x, np.sin(x))

ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.set_title('Sine Wave — OO Style')
plt.show()
````

### Functional vs OO — Side by Side

| Functional (`plt.*`)        | Object-Oriented (`ax.*`)       |
|-----------------------------|--------------------------------|
| `plt.plot(x, y)`           | `ax.plot(x, y)`                |
| `plt.xlabel('X')`          | `ax.set_xlabel('X')`           |
| `plt.ylabel('Y')`          | `ax.set_ylabel('Y')`           |
| `plt.title('Title')`       | `ax.set_title('Title')`        |
| `plt.xlim(0, 10)`          | `ax.set_xlim(0, 10)`           |
| `plt.legend()`             | `ax.legend()`                  |

## Method 2: `fig.add_axes()` — Manual Axes Placement

`fig.add_axes([left, bottom, width, height])` lets you position axes anywhere on the figure using fractional coordinates (0 to 1). This is useful for inset plots or custom layouts.

````python
fig = plt.figure()

# Main axes: starts at 10% from left, 10% from bottom, 80% wide, 80% tall
ax_main = fig.add_axes([0.1, 0.1, 0.8, 0.8])

x = np.linspace(0, 10, 200)
ax_main.plot(x, np.sin(x), 'b-')
ax_main.set_xlabel('x')
ax_main.set_ylabel('sin(x)')
ax_main.set_title('Main Plot with Inset')

# Inset axes: positioned inside the main plot
ax_inset = fig.add_axes([0.55, 0.55, 0.3, 0.3])
ax_inset.plot(x, np.cos(x), 'r-')
ax_inset.set_title('Inset: cos(x)', fontsize=9)

plt.show()
````

````python
# Two axes side by side using add_axes
fig = plt.figure()

ax_left = fig.add_axes([0.1, 0.1, 0.35, 0.8])
ax_right = fig.add_axes([0.55, 0.1, 0.35, 0.8])

x = np.linspace(0, 5, 100)

ax_left.plot(x, x ** 2, 'b')
ax_left.set_title('Quadratic')
ax_left.set_xlabel('x')
ax_left.set_ylabel('x²')

ax_right.plot(x, np.sqrt(x), 'r')
ax_right.set_title('Square Root')
ax_right.set_xlabel('x')
ax_right.set_ylabel('√x')

plt.show()
````

## Setting Label, Title, and Limits in One Call

The `ax.set()` method accepts keyword arguments to configure multiple properties at once, reducing boilerplate.

````python
fig, ax = plt.subplots()

x = np.linspace(0, 10, 200)
ax.plot(x, np.exp(-x / 3) * np.cos(2 * x))

# Set multiple properties at once
ax.set(
    xlabel='Time (s)',
    ylabel='Amplitude',
    title='Damped Cosine',
    xlim=(0, 10),
    ylim=(-1, 1)
)
plt.show()
````

## Figure Size with `figsize`

The `figsize` parameter takes a `(width, height)` tuple in **inches**. The default is `(6.4, 4.8)`.

Choosing the right figure size matters for:
- Readability in notebooks vs printed reports
- Aspect ratio appropriate for the data
- Fitting into slide layouts

````python
# Wide figure — good for time series
fig, ax = plt.subplots(figsize=(12, 3))

x = np.linspace(0, 50, 1000)
y = np.sin(x) + 0.3 * np.sin(5 * x)

ax.plot(x, y)
ax.set_title('Wide Figure (12 x 3 inches)')
ax.set_xlabel('x')
plt.show()
````

````python
# Square figure — good for scatter plots or circular data
fig, ax = plt.subplots(figsize=(5, 5))

theta = np.linspace(0, 2 * np.pi, 100)
ax.plot(np.cos(theta), np.sin(theta))
ax.set_title('Square Figure (5 x 5 inches)')
ax.set_aspect('equal')  # Ensure the circle looks circular
plt.show()
````

## DPI — Dots Per Inch

DPI controls the resolution of the rendered image:
- **Screen display:** 72-100 DPI is typical.
- **Print quality:** 150-300 DPI is standard.
- **Publication:** 300+ DPI is often required.

DPI can be set on the figure (for display) or in `savefig()` (for export).

````python
# Compare different DPI settings
x = np.linspace(0, 2 * np.pi, 200)

fig_low, ax_low = plt.subplots(figsize=(4, 3), dpi=50)
ax_low.plot(x, np.sin(x))
ax_low.set_title('Low DPI (50) — Pixelated')
plt.show()

fig_high, ax_high = plt.subplots(figsize=(4, 3), dpi=150)
ax_high.plot(x, np.sin(x))
ax_high.set_title('High DPI (150) — Crisp')
plt.show()
````

## `fig.tight_layout()`

When figures have multiple axes or long labels, elements can overlap. `fig.tight_layout()` automatically adjusts spacing to prevent this. Call it after all plotting is done, just before `plt.show()` or `plt.savefig()`.

````python
# Without tight_layout — labels may overlap
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

x = np.linspace(0, 10, 100)

ax1.plot(x, np.sin(x))
ax1.set_title('Sine Function')
ax1.set_xlabel('Horizontal Axis Label')
ax1.set_ylabel('Vertical Axis Label')

ax2.plot(x, np.cos(x))
ax2.set_title('Cosine Function')
ax2.set_xlabel('Horizontal Axis Label')
ax2.set_ylabel('Vertical Axis Label')

# Try commenting out the next line to see the overlap
fig.tight_layout()
plt.show()
````

## Comparing the Two Interfaces

Below is the same plot created both ways. The OO approach is slightly longer but much more explicit about which axes you are drawing on.

````python
x = np.linspace(0, 10, 200)
y1 = np.sin(x)
y2 = np.cos(x)

# --- Functional approach ---
plt.figure(figsize=(8, 3))
plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Functional Approach')
plt.legend()
plt.show()
````

````python
# --- Object-oriented approach ---
fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Object-Oriented Approach')
ax.legend()
plt.show()
````

### When to Use Which

| Scenario | Recommended |
|----------|-------------|
| Quick exploratory plot | Functional (`plt.*`) |
| Multiple subplots | Object-Oriented |
| Custom layouts | Object-Oriented |
| Embedding in GUI apps | Object-Oriented |
| Production/publication code | Object-Oriented |
| One-liner in a REPL | Functional |

## Saving Figures (OO Style)

With the OO interface, call `fig.savefig()` on the figure object rather than `plt.savefig()`.

````python
fig, ax = plt.subplots(figsize=(6, 4))

x = np.linspace(-5, 5, 300)
ax.plot(x, np.tanh(x), 'b-', linewidth=2)
ax.set(xlabel='x', ylabel='tanh(x)', title='Hyperbolic Tangent')
ax.axhline(y=0, color='gray', linewidth=0.5)  # Reference line at y=0
ax.axvline(x=0, color='gray', linewidth=0.5)  # Reference line at x=0

fig.savefig('data/tanh_plot.png', dpi=150, bbox_inches='tight')
plt.show()

print('Saved to data/tanh_plot.png')
````

## Practical Example: Multi-Panel Comparison

Combining figure size, DPI, tight_layout, and the OO interface for a polished visualization.

````python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

x = np.linspace(0, 4 * np.pi, 300)

# Plot 1: Sine variations
axes[0].plot(x, np.sin(x), label='sin(x)')
axes[0].plot(x, np.sin(2 * x), label='sin(2x)')
axes[0].set_title('Frequency')
axes[0].legend(fontsize=9)

# Plot 2: Amplitude variations
axes[1].plot(x, np.sin(x), label='A=1')
axes[1].plot(x, 2 * np.sin(x), label='A=2')
axes[1].set_title('Amplitude')
axes[1].legend(fontsize=9)

# Plot 3: Phase variations
axes[2].plot(x, np.sin(x), label='phase=0')
axes[2].plot(x, np.sin(x + np.pi / 2), label='phase=pi/2')
axes[2].set_title('Phase Shift')
axes[2].legend(fontsize=9)

for ax in axes:
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')

fig.suptitle('Properties of Sine Waves', fontsize=14, y=1.02)
fig.tight_layout()
plt.show()
````

## Summary

| Concept | Key API |
|---------|---------|
| Create figure + axes | `fig, ax = plt.subplots()` |
| Manual axes placement | `fig.add_axes([left, bottom, w, h])` |
| Labels and title (OO) | `ax.set_xlabel()`, `ax.set_ylabel()`, `ax.set_title()` |
| Batch-set properties | `ax.set(xlabel=..., ylabel=..., title=...)` |
| Figure size | `plt.subplots(figsize=(w, h))` |
| Resolution | `plt.subplots(dpi=...)` or `fig.savefig(..., dpi=...)` |
| Prevent label overlap | `fig.tight_layout()` |
| Save (OO) | `fig.savefig('path.png')` |
