# Subplots and Multi-Panel Layouts

Real-world data analysis often requires comparing several plots side by side. This notebook covers the various ways Matplotlib lets you arrange multiple axes within a single figure — from simple grids to complex custom layouts.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Basic Grid with `plt.subplots(nrows, ncols)`

Pass the number of rows and columns to `plt.subplots()` to create a uniform grid of axes.

```python
# 1 row, 2 columns
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

x = np.linspace(0, 10, 200)

axes[0].plot(x, np.sin(x), 'b')
axes[0].set_title('sin(x)')

axes[1].plot(x, np.cos(x), 'r')
axes[1].set_title('cos(x)')

fig.tight_layout()
plt.show()
```

```python
# 2 rows, 1 column — stacked vertically
fig, axes = plt.subplots(2, 1, figsize=(8, 6))

t = np.linspace(0, 1, 500)

axes[0].plot(t, np.sin(2 * np.pi * 5 * t))
axes[0].set_title('5 Hz Sine Wave')
axes[0].set_ylabel('Amplitude')

axes[1].plot(t, np.sin(2 * np.pi * 20 * t), 'r')
axes[1].set_title('20 Hz Sine Wave')
axes[1].set_xlabel('Time (s)')
axes[1].set_ylabel('Amplitude')

fig.tight_layout()
plt.show()
```

## Accessing Axes in a 2D Grid

When both `nrows > 1` and `ncols > 1`, `axes` becomes a 2D NumPy array. Access individual axes with `axes[row, col]`.

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 6))

x = np.linspace(0, 2 * np.pi, 200)

# Top row
axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title('sin(x)')

axes[0, 1].plot(x, np.sin(2 * x))
axes[0, 1].set_title('sin(2x)')

axes[0, 2].plot(x, np.sin(3 * x))
axes[0, 2].set_title('sin(3x)')

# Bottom row
axes[1, 0].plot(x, np.cos(x), 'r')
axes[1, 0].set_title('cos(x)')

axes[1, 1].plot(x, np.cos(2 * x), 'r')
axes[1, 1].set_title('cos(2x)')

axes[1, 2].plot(x, np.cos(3 * x), 'r')
axes[1, 2].set_title('cos(3x)')

fig.tight_layout()
plt.show()
```

### Flattening the Axes Array

When iterating over all axes, it is convenient to flatten the 2D array into 1D.

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 6))

x = np.linspace(0, 10, 200)
functions = [
    ('x', x),
    ('x²', x ** 2),
    ('x³', x ** 3),
    ('sqrt(x)', np.sqrt(x)),
    ('log(1+x)', np.log1p(x)),
    ('exp(x/5)', np.exp(x / 5)),
]

for ax, (name, y) in zip(axes.flat, functions):
    ax.plot(x, y)
    ax.set_title(name)

fig.tight_layout()
plt.show()
```

## `fig.add_subplot()` — Adding Axes One at a Time

An alternative approach: create the figure first, then add subplots with `fig.add_subplot(nrows, ncols, index)`. The index starts at 1 (not 0), counting left-to-right, top-to-bottom.

```python
fig = plt.figure(figsize=(10, 4))

x = np.linspace(0, 5, 100)

# add_subplot(nrows, ncols, index)
ax1 = fig.add_subplot(1, 3, 1)  # Row 1, Col 1
ax1.plot(x, x)
ax1.set_title('Linear')

ax2 = fig.add_subplot(1, 3, 2)  # Row 1, Col 2
ax2.plot(x, x ** 2)
ax2.set_title('Quadratic')

ax3 = fig.add_subplot(1, 3, 3)  # Row 1, Col 3
ax3.plot(x, x ** 3)
ax3.set_title('Cubic')

fig.tight_layout()
plt.show()
```

## Sharing Axes

When comparing plots, shared axes make it easier to see differences. Use `sharex` and/or `sharey` to link axis limits across subplots.

```python
np.random.seed(0)

# Without shared y-axis — hard to compare magnitudes
fig, axes = plt.subplots(1, 3, figsize=(12, 3))

for i, ax in enumerate(axes):
    data = np.random.randn(100) * (i + 1)  # Different scales
    ax.hist(data, bins=20)
    ax.set_title(f'std = {i + 1}')

fig.suptitle('Without sharey — Different Scales', fontsize=12)
fig.tight_layout()
plt.show()
```

```python
np.random.seed(0)

# With shared y-axis — easy to compare
fig, axes = plt.subplots(1, 3, figsize=(12, 3), sharey=True)

for i, ax in enumerate(axes):
    data = np.random.randn(100) * (i + 1)
    ax.hist(data, bins=20)
    ax.set_title(f'std = {i + 1}')

fig.suptitle('With sharey — Same Scale', fontsize=12)
fig.tight_layout()
plt.show()
```

```python
# Shared x-axis is useful for stacked time series
t = np.linspace(0, 10, 500)

fig, axes = plt.subplots(3, 1, figsize=(10, 6), sharex=True)

axes[0].plot(t, np.sin(t))
axes[0].set_ylabel('sin(t)')

axes[1].plot(t, np.sin(3 * t), 'r')
axes[1].set_ylabel('sin(3t)')

axes[2].plot(t, np.sin(t) + np.sin(3 * t), 'g')
axes[2].set_ylabel('sum')
axes[2].set_xlabel('Time (s)')

fig.suptitle('Signal Decomposition', fontsize=13)
fig.tight_layout()
plt.show()
```

## `plt.GridSpec` — Complex Layouts

When you need axes of different sizes (e.g., a large main plot with smaller side panels), `GridSpec` lets you define a grid and have individual axes span multiple cells.

```python
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(10, 6))
gs = GridSpec(2, 3, figure=fig)  # 2 rows, 3 columns

# Large plot spanning the left two columns, both rows
ax_main = fig.add_subplot(gs[:, :2])

# Two small plots on the right
ax_top_right = fig.add_subplot(gs[0, 2])
ax_bottom_right = fig.add_subplot(gs[1, 2])

# Generate sample data
x = np.linspace(0, 10, 200)

ax_main.plot(x, np.sin(x) * np.exp(-x / 5), 'b-', linewidth=2)
ax_main.set_title('Main: Damped Sine', fontsize=13)
ax_main.set_xlabel('x')
ax_main.set_ylabel('f(x)')

ax_top_right.plot(x, np.sin(x), 'r-')
ax_top_right.set_title('sin(x)', fontsize=10)

ax_bottom_right.plot(x, np.exp(-x / 5), 'g-')
ax_bottom_right.set_title('exp(-x/5)', fontsize=10)

fig.tight_layout()
plt.show()
```

```python
# Another GridSpec example: scatter plot with marginal histograms
np.random.seed(42)
data_x = np.random.randn(300)
data_y = data_x * 0.5 + np.random.randn(300) * 0.5

fig = plt.figure(figsize=(7, 7))
gs = GridSpec(4, 4, figure=fig, hspace=0.05, wspace=0.05)

# Main scatter plot
ax_scatter = fig.add_subplot(gs[1:, :3])
ax_scatter.scatter(data_x, data_y, alpha=0.5, s=15)
ax_scatter.set_xlabel('X')
ax_scatter.set_ylabel('Y')

# Top histogram (marginal distribution of X)
ax_hist_top = fig.add_subplot(gs[0, :3], sharex=ax_scatter)
ax_hist_top.hist(data_x, bins=30, color='steelblue', edgecolor='white')
ax_hist_top.tick_params(labelbottom=False)
ax_hist_top.set_ylabel('Count')

# Right histogram (marginal distribution of Y)
ax_hist_right = fig.add_subplot(gs[1:, 3], sharey=ax_scatter)
ax_hist_right.hist(data_y, bins=30, orientation='horizontal', color='salmon', edgecolor='white')
ax_hist_right.tick_params(labelleft=False)
ax_hist_right.set_xlabel('Count')

fig.suptitle('Scatter with Marginal Histograms', fontsize=13, y=0.98)
plt.show()
```

## Adjusting Spacing

Several mechanisms control whitespace between subplots:

| Method | Description |
|--------|-------------|
| `fig.tight_layout()` | Automatic spacing — works well in most cases |
| `fig.subplots_adjust(...)` | Manual control: `left`, `right`, `top`, `bottom`, `wspace`, `hspace` |
| `GridSpec(hspace=..., wspace=...)` | Spacing built into the grid specification |
| `constrained_layout=True` | Alternative to `tight_layout`, set at figure creation |

```python
# subplots_adjust — manual spacing control
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

x = np.linspace(0, 5, 100)
for i, ax in enumerate(axes.flat):
    ax.plot(x, x ** (i + 1))
    ax.set_title(f'x^{i + 1}')

# wspace = horizontal gap, hspace = vertical gap (fraction of axes size)
fig.subplots_adjust(wspace=0.4, hspace=0.5)
plt.show()
```

```python
# constrained_layout — modern alternative to tight_layout
fig, axes = plt.subplots(2, 2, figsize=(8, 6), constrained_layout=True)

x = np.linspace(0, 5, 100)
for i, ax in enumerate(axes.flat):
    ax.plot(x, np.sin((i + 1) * x))
    ax.set_title(f'sin({i + 1}x)')
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')

plt.show()
```

## Removing or Hiding Unused Axes

When your data does not fill the entire grid, hide the empty axes to keep the figure clean.

```python
# 5 plots in a 2x3 grid — one axes is empty
fig, axes = plt.subplots(2, 3, figsize=(12, 6))

x = np.linspace(0, 5, 100)
functions = ['sin(x)', 'cos(x)', 'tan(x)', 'exp(-x)', 'log(1+x)']
values = [np.sin(x), np.cos(x), np.clip(np.tan(x), -5, 5), np.exp(-x), np.log1p(x)]

for ax, name, y in zip(axes.flat, functions, values):
    ax.plot(x, y)
    ax.set_title(name)

# Hide the last (empty) axes
axes[1, 2].set_visible(False)

fig.tight_layout()
plt.show()
```

## Practical Example: Dashboard Layout

Using GridSpec to build a data dashboard with different sized panels.

```python
np.random.seed(7)

fig = plt.figure(figsize=(12, 7))
gs = GridSpec(3, 3, figure=fig)

# Top: wide time series spanning all 3 columns
ax_ts = fig.add_subplot(gs[0, :])
t = np.arange(200)
signal = np.cumsum(np.random.randn(200))
ax_ts.plot(t, signal, color='steelblue')
ax_ts.set_title('Time Series (Random Walk)')
ax_ts.set_xlabel('Step')

# Middle-left: histogram (2 cols wide)
ax_hist = fig.add_subplot(gs[1, :2])
ax_hist.hist(np.random.randn(500), bins=30, color='coral', edgecolor='white')
ax_hist.set_title('Distribution')

# Middle-right: box plot
ax_box = fig.add_subplot(gs[1, 2])
data = [np.random.randn(50) + i for i in range(4)]
ax_box.boxplot(data)
ax_box.set_title('Box Plot')

# Bottom-left: scatter
ax_scatter = fig.add_subplot(gs[2, 0])
ax_scatter.scatter(np.random.randn(50), np.random.randn(50), s=20, alpha=0.7)
ax_scatter.set_title('Scatter')

# Bottom-center + right: bar chart (2 cols wide)
ax_bar = fig.add_subplot(gs[2, 1:])
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 100, size=5)
ax_bar.bar(categories, values, color='mediumseagreen')
ax_bar.set_title('Bar Chart')

fig.suptitle('Dashboard Layout with GridSpec', fontsize=14, fontweight='bold')
fig.tight_layout()
plt.show()
```

## Summary

| Task | API |
|------|-----|
| Uniform grid | `fig, axes = plt.subplots(nrows, ncols)` |
| Access axes (1D) | `axes[i]` |
| Access axes (2D grid) | `axes[row, col]` |
| Iterate all axes | `for ax in axes.flat:` |
| Add axes one at a time | `fig.add_subplot(nrows, ncols, index)` |
| Share axis limits | `plt.subplots(..., sharex=True, sharey=True)` |
| Complex layouts | `GridSpec(nrows, ncols)` + `fig.add_subplot(gs[slice])` |
| Auto spacing | `fig.tight_layout()` or `constrained_layout=True` |
| Manual spacing | `fig.subplots_adjust(wspace=..., hspace=...)` |
| Hide empty axes | `ax.set_visible(False)` |
