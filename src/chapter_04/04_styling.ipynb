# Styling Plots

Default Matplotlib plots are functional but often need visual polish for presentations, reports, or publications. This notebook covers the full range of styling options — from individual line properties to global themes.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Line Styles

The `linestyle` (or `ls`) parameter controls how the line is drawn between data points.

```python
x = np.linspace(0, 10, 200)
fig, ax = plt.subplots(figsize=(8, 4))

styles = [
    ('solid', '-'),
    ('dashed', '--'),
    ('dash-dot', '-.'),
    ('dotted', ':'),
]

for i, (name, style) in enumerate(styles):
    ax.plot(x, np.sin(x + i * 0.5), linestyle=style, linewidth=2,
            label=f'{name} ({style})')

ax.legend(fontsize=10)
ax.set_title('Line Styles')
ax.set_xlabel('x')
plt.show()
```

## Line Width

The `linewidth` (or `lw`) parameter controls thickness. Default is 1.5.

```python
x = np.linspace(0, 5, 200)
fig, ax = plt.subplots(figsize=(8, 4))

for lw in [0.5, 1, 2, 3, 5]:
    ax.plot(x, np.sin(x) + lw, linewidth=lw, label=f'lw = {lw}')

ax.legend()
ax.set_title('Line Width Comparison')
plt.show()
```

## Color Specification

Matplotlib accepts colors in many formats:

| Format | Example | Description |
|--------|---------|-------------|
| Single letter | `'r'` | Short color codes (r, g, b, c, m, y, k, w) |
| Named color | `'steelblue'` | Any CSS4 named color |
| Hex string | `'#2ca02c'` | Standard hex RGB |
| RGB tuple | `(0.1, 0.5, 0.8)` | Float values 0-1 |
| RGBA tuple | `(0.1, 0.5, 0.8, 0.5)` | With alpha (transparency) |
| Grayscale string | `'0.6'` | Float as a string, 0=black, 1=white |

```python
x = np.linspace(0, 3, 100)
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(x, x, color='r', linewidth=2, label="'r' (short code)")
ax.plot(x, x + 0.5, color='steelblue', linewidth=2, label="'steelblue' (named)")
ax.plot(x, x + 1.0, color='#e377c2', linewidth=2, label="'#e377c2' (hex)")
ax.plot(x, x + 1.5, color=(0.2, 0.7, 0.3), linewidth=2, label='(0.2, 0.7, 0.3) (RGB)')
ax.plot(x, x + 2.0, color='0.5', linewidth=2, label="'0.5' (grayscale)")

ax.legend(loc='upper left', fontsize=9)
ax.set_title('Color Specification Methods')
plt.show()
```

## Marker Styles

Markers highlight individual data points along a line (or can be used alone for scatter-like plots).

```python
x = np.linspace(0, 5, 15)
fig, ax = plt.subplots(figsize=(10, 6))

markers = ['o', 's', '^', 'D', 'v', 'p', '*', 'h', 'X', '+']
names = ['circle', 'square', 'triangle up', 'diamond', 'triangle down',
         'pentagon', 'star', 'hexagon', 'X', 'plus']

for i, (marker, name) in enumerate(zip(markers, names)):
    ax.plot(x, np.sin(x) + i * 0.4, marker=marker, markersize=8,
            label=f"{marker} — {name}", linewidth=1)

ax.legend(loc='upper left', fontsize=8, ncol=2)
ax.set_title('Common Marker Styles')
plt.show()
```

```python
# Marker customization: face color, edge color, edge width
x = np.linspace(0, 5, 12)
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, np.sin(x), 'o-',
        markersize=12,
        markerfacecolor='gold',
        markeredgecolor='black',
        markeredgewidth=1.5,
        linewidth=1.5,
        color='gray',
        label='Custom markers')

ax.legend()
ax.set_title('Custom Marker Appearance')
plt.show()
```

## Built-in Style Sheets

Matplotlib ships with many pre-defined style sheets that change the overall look of plots. Use `plt.style.use()` to apply one globally, or `plt.style.context()` to apply one temporarily.

```python
# List all available styles
print('Available styles:')
print(sorted(plt.style.available))
```

```python
# Compare several styles side by side
styles_to_show = ['default', 'ggplot', 'seaborn-v0_8', 'dark_background']

x = np.linspace(0, 2 * np.pi, 100)

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for ax, style_name in zip(axes.flat, styles_to_show):
    with plt.style.context(style_name):
        # We need to plot on the pre-created axes, so manually
        # apply some style properties
        ax.plot(x, np.sin(x), label='sin')
        ax.plot(x, np.cos(x), label='cos')
        ax.set_title(f"Style: '{style_name}'", fontsize=11)
        ax.legend(fontsize=9)

fig.tight_layout()
plt.show()
```

```python
# Applying a style temporarily with context manager
x = np.linspace(0, 10, 200)

with plt.style.context('ggplot'):
    fig, ax = plt.subplots(figsize=(8, 4))
    ax.plot(x, np.sin(x), label='sin(x)')
    ax.plot(x, np.cos(x), label='cos(x)')
    ax.set_title('Using ggplot Style (via context manager)')
    ax.legend()
    plt.show()
```

## Legend Placement

The `loc` parameter controls where the legend appears. Pass a string or a numeric code.

```python
locations = [
    'upper left', 'upper center', 'upper right',
    'center left', 'center', 'center right',
    'lower left', 'lower center', 'lower right',
]

fig, axes = plt.subplots(3, 3, figsize=(14, 10))
x = np.linspace(0, 5, 50)

for ax, loc in zip(axes.flat, locations):
    ax.plot(x, np.sin(x), label='sin(x)')
    ax.plot(x, np.cos(x), label='cos(x)')
    ax.legend(loc=loc)
    ax.set_title(f"loc='{loc}'", fontsize=9)

fig.tight_layout()
plt.show()
```

```python
# Place legend outside the plot using bbox_to_anchor
fig, ax = plt.subplots(figsize=(8, 4))

x = np.linspace(0, 10, 100)
for i in range(5):
    ax.plot(x, np.sin(x + i * 0.5), label=f'Phase {i}')

# Move legend to the right of the plot
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
ax.set_title('Legend Outside the Plot')
fig.tight_layout()
plt.show()
```

## Grid Lines

Grid lines help readers trace values across the plot area.

```python
x = np.linspace(0, 10, 100)

fig, axes = plt.subplots(1, 3, figsize=(14, 4))

# Default grid
axes[0].plot(x, np.sin(x))
axes[0].grid(True)
axes[0].set_title('Default Grid')

# Custom grid
axes[1].plot(x, np.sin(x))
axes[1].grid(True, linestyle='--', alpha=0.5, color='gray')
axes[1].set_title('Styled Grid')

# Major and minor grid
axes[2].plot(x, np.sin(x))
axes[2].grid(True, which='major', linewidth=1, alpha=0.7)
axes[2].grid(True, which='minor', linewidth=0.5, alpha=0.3)
axes[2].minorticks_on()
axes[2].set_title('Major + Minor Grid')

fig.tight_layout()
plt.show()
```

## Annotations

`ax.annotate()` adds text with an optional arrow pointing to a specific data point. This is useful for highlighting outliers, peaks, or interesting features.

```python
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y, 'b-', linewidth=2)

# Annotate the peak
ax.annotate(
    'Maximum',                        # Text
    xy=(np.pi / 2, 1),               # Point to annotate
    xytext=(np.pi / 2 + 1, 0.7),     # Text position
    fontsize=12,
    arrowprops=dict(
        arrowstyle='->',
        color='red',
        linewidth=1.5
    )
)

# Annotate the minimum
ax.annotate(
    'Minimum',
    xy=(3 * np.pi / 2, -1),
    xytext=(3 * np.pi / 2 - 1.5, -0.5),
    fontsize=12,
    arrowprops=dict(
        arrowstyle='->',
        color='red',
        linewidth=1.5
    )
)

ax.set_title('Annotated Sine Wave')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
plt.show()
```

```python
# Different arrow styles
fig, ax = plt.subplots(figsize=(8, 5))

arrow_styles = ['->', '-|>', 'fancy', 'wedge', '-[']

for i, style in enumerate(arrow_styles):
    ax.annotate(
        f'arrowstyle="{style}"',
        xy=(0.3, 0.8 - i * 0.15),
        xytext=(0.6, 0.8 - i * 0.15),
        fontsize=10,
        arrowprops=dict(arrowstyle=style, color='steelblue', linewidth=1.5),
        va='center',
        transform=ax.transAxes,
    )

ax.set_title('Arrow Styles for Annotations')
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
plt.show()
```

## Adding Text

`ax.text()` places text at arbitrary data coordinates. Unlike `annotate()`, it does not draw arrows.

```python
fig, ax = plt.subplots(figsize=(8, 5))

x = np.linspace(0, 10, 200)
ax.plot(x, np.exp(-x / 3), 'b-', linewidth=2)

# Place text at specific data coordinates
ax.text(2, 0.5, r'$f(x) = e^{-x/3}$', fontsize=16, color='darkblue',
        bbox=dict(boxstyle='round,pad=0.3', facecolor='lightyellow',
                  edgecolor='gray', alpha=0.8))

# Text in figure coordinates using transform
ax.text(0.95, 0.95, 'Exponential\nDecay', transform=ax.transAxes,
        fontsize=11, verticalalignment='top', horizontalalignment='right',
        style='italic', color='gray')

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Adding Text to Plots')
plt.show()
```

## Colorbars

Colorbars map numeric values to colors. They are essential for any plot that encodes data as color (scatter with color-mapped points, heatmaps, contour plots, etc.).

```python
# Scatter plot with color-mapped values
np.random.seed(42)
x = np.random.randn(200)
y = np.random.randn(200)
colors = np.sqrt(x ** 2 + y ** 2)  # Distance from origin

fig, ax = plt.subplots(figsize=(7, 5))
scatter = ax.scatter(x, y, c=colors, cmap='viridis', s=50, alpha=0.7)

cbar = fig.colorbar(scatter, ax=ax)
cbar.set_label('Distance from Origin')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Scatter with Colorbar')
plt.show()
```

```python
# Common colormaps
fig, axes = plt.subplots(2, 3, figsize=(14, 8))

cmaps = ['viridis', 'plasma', 'coolwarm', 'RdYlGn', 'magma', 'cividis']

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

for ax, cmap_name in zip(axes.flat, cmaps):
    im = ax.pcolormesh(X, Y, Z, cmap=cmap_name, shading='auto')
    fig.colorbar(im, ax=ax, shrink=0.8)
    ax.set_title(cmap_name)
    ax.set_aspect('equal')

fig.tight_layout()
plt.show()
```

### Choosing a Colormap

- **Sequential** (viridis, plasma, magma, cividis): single-direction data (counts, intensity).
- **Diverging** (coolwarm, RdYlGn, seismic): data with a meaningful midpoint (temperature anomaly, correlation).
- **Qualitative** (tab10, Set2, Paired): categorical data.

**Tip:** viridis is perceptually uniform and colorblind-friendly — it is a safe default.

## Transparency (Alpha)

The `alpha` parameter (0=fully transparent, 1=fully opaque) is useful when points or lines overlap.

```python
np.random.seed(10)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

x = np.random.randn(500)
y = np.random.randn(500)

axes[0].scatter(x, y, s=30)
axes[0].set_title('Without alpha — Overlap hidden')

axes[1].scatter(x, y, s=30, alpha=0.3)
axes[1].set_title('With alpha=0.3 — Density visible')

fig.tight_layout()
plt.show()
```

## Putting It All Together

A polished publication-style plot combining multiple styling techniques.

```python
np.random.seed(99)

fig, ax = plt.subplots(figsize=(9, 5))

# Data
x = np.linspace(0, 10, 50)
y_true = 2 * np.sin(x)
y_noisy = y_true + np.random.randn(50) * 0.5

# Noisy observations
ax.scatter(x, y_noisy, color='steelblue', s=30, alpha=0.6,
           label='Observations', zorder=3)

# True function
ax.plot(x, y_true, color='coral', linewidth=2.5, linestyle='-',
        label='True signal: 2sin(x)')

# Confidence band
ax.fill_between(x, y_true - 0.5, y_true + 0.5, alpha=0.15, color='coral',
                label='Confidence band')

# Annotation
ax.annotate('Peak', xy=(np.pi / 2, 2), xytext=(2.5, 2.8),
            arrowprops=dict(arrowstyle='->', color='black'),
            fontsize=11, fontweight='bold')

# Styling
ax.set_xlabel('Time (s)', fontsize=12)
ax.set_ylabel('Amplitude', fontsize=12)
ax.set_title('Signal Recovery from Noisy Data', fontsize=14)
ax.legend(loc='lower left', fontsize=10, framealpha=0.9)
ax.grid(True, linestyle='--', alpha=0.3)
ax.set_xlim(0, 10)

fig.tight_layout()
plt.show()
```

## Summary

| Property | Parameter | Example Values |

| Line style | `linestyle` / `ls` | `'-'`, `'--'`, `'-.'`, `':'` |

| Color | `color` / `c` | `'red'`, `'#ff0000'`, `(1,0,0)` |
| Marker shape | `marker` | `'o'`, `'s'`, `'^'`, `'D'` |
| Marker size | `markersize` / `ms` | `5`, `10`, `15` |
| Transparency | `alpha` | `0.0` to `1.0` |
| Style sheet | `plt.style.use()` | `'ggplot'`, `'seaborn-v0_8'` |
| Grid | `ax.grid()` | `True/False`, `which`, `linestyle` |
| Annotation | `ax.annotate()` | Text + arrow to data point |
| Text | `ax.text()` | Free-form text at coordinates |
| Colorbar | `fig.colorbar()` | Maps color to numeric value |
