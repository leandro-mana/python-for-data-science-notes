# Matplotlib Basics

Matplotlib is the foundational plotting library in the Python data science ecosystem. Nearly every other visualization library (Seaborn, Pandas plotting, etc.) is built on top of it. This notebook covers the essential functional interface — `pyplot` — which provides a MATLAB-like way to create plots quickly.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## The `%matplotlib inline` Magic

The `%matplotlib inline` magic command tells Jupyter to render plots directly inside the notebook output cells rather than in a separate window. This is the standard approach when working in notebooks.

Other backends include:
- `%matplotlib notebook` — interactive plots with zoom/pan (classic Jupyter)
- `%matplotlib widget` — interactive plots in JupyterLab
- `%matplotlib qt` — external Qt window

## Creating a Simple Line Plot

`plt.plot()` is the most basic plotting function. Pass it a sequence of values and it draws a line connecting them.

```python
# When only y-values are provided, the x-axis defaults to 0, 1, 2, ...
plt.plot([2, 4, 1, 5, 3])
plt.show()
```

## Providing Both x and y Values

For more control, pass both x and y arrays. Using NumPy arrays is the standard approach since they support vectorized operations.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

## `plt.show()` Behavior

- In a **script** (`.py` file), `plt.show()` is required to display the plot — it opens the GUI window and blocks execution until the window is closed.
- In a **Jupyter notebook** with `%matplotlib inline`, plots render automatically at the end of a cell. Calling `plt.show()` is optional but considered good practice as it suppresses the text representation of the plot object.
- Each call to `plt.show()` finalizes the current figure. Subsequent plotting commands start a new figure.

```python
# Without plt.show(), the notebook still renders the plot
# but may also print a text representation like <matplotlib.lines.Line2D ...>
x = np.linspace(0, 5, 50)
plt.plot(x, x ** 2)
```

```python
# Using plt.show() gives a cleaner output
plt.plot(x, x ** 2)
plt.show()
```

## Adding Labels and a Title

Unlabeled plots are nearly useless in practice. Always add axis labels and a title to communicate what the data represents.

```python
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('Angle (radians)')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.show()
```

## Controlling Axis Limits

By default, Matplotlib auto-scales axes to fit the data. Use `plt.xlim()` and `plt.ylim()` to override this and zoom in on a region of interest.

```python
x = np.linspace(0, 4 * np.pi, 300)
y = np.sin(x)

plt.plot(x, y)
plt.xlim(0, 2 * np.pi)   # Show only the first full cycle
plt.ylim(-1.5, 1.5)      # Add some vertical padding
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('First Period of Sine Wave')
plt.show()
```

```python
# plt.axis() can set both limits at once: [xmin, xmax, ymin, ymax]
plt.plot(x, y)
plt.axis([0, np.pi, 0, 1.2])  # Zoom into the first positive half-cycle
plt.title('Zoomed View with plt.axis()')
plt.show()
```

## Plotting Multiple Lines

Call `plt.plot()` multiple times before `plt.show()` to overlay several lines on the same axes. Matplotlib automatically cycles through a default color palette.

```python
x = np.linspace(0, 2 * np.pi, 200)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.plot(x, np.sin(x) + np.cos(x))

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Trigonometric Functions')
plt.show()
```

## Adding a Legend

When a plot has multiple lines, a legend is essential. There are two common approaches:

1. Pass a `label` argument to each `plt.plot()` call, then call `plt.legend()`.
2. Pass a list of labels directly to `plt.legend()`.

```python
# Approach 1: label parameter (preferred — keeps label near data definition)
x = np.linspace(0, 2 * np.pi, 200)

plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Trig Functions with Legend')
plt.legend()
plt.show()
```

```python
# Approach 2: pass labels directly to plt.legend()
x = np.linspace(0, 5, 100)

plt.plot(x, x)
plt.plot(x, x ** 2)
plt.plot(x, x ** 3)

plt.legend(['Linear', 'Quadratic', 'Cubic'])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Polynomial Growth')
plt.show()
```

## Plotting with NumPy-Generated Data

In practice, you will rarely type out data by hand. NumPy provides many functions for generating arrays that make good sample data for learning to plot.

```python
# Random walk example
np.random.seed(42)
steps = np.random.choice([-1, 1], size=500)
walk = np.cumsum(steps)

plt.plot(walk)
plt.xlabel('Step Number')
plt.ylabel('Position')
plt.title('1D Random Walk')
plt.show()
```

```python
# Exponential decay
t = np.linspace(0, 5, 200)
for tau in [0.5, 1.0, 2.0]:
    plt.plot(t, np.exp(-t / tau), label=f'tau = {tau}')

plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Exponential Decay')
plt.legend()
plt.show()
```

## Saving Figures with `plt.savefig()`

Use `plt.savefig()` to export a plot to a file. The format is inferred from the file extension. Common formats:

| Format | Extension | Notes |
|--------|-----------|-------|
| PNG    | `.png`    | Raster, good for web/presentations |
| PDF    | `.pdf`    | Vector, ideal for publications |
| SVG    | `.svg`    | Vector, good for web |
| JPEG   | `.jpg`    | Raster, lossy compression |

**Important:** Call `plt.savefig()` *before* `plt.show()`, because `plt.show()` clears the current figure.

```python
x = np.linspace(0, 10, 200)
y = np.sin(x) * np.exp(-x / 5)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Damped Sine')
plt.title('Damped Oscillation')

# Save before showing
plt.savefig('data/damped_oscillation.png', dpi=150, bbox_inches='tight')
plt.show()

print('Figure saved to data/damped_oscillation.png')
```

### `savefig()` Key Parameters

- `dpi` — Dots per inch; higher values produce larger, sharper images (default is 100).
- `bbox_inches='tight'` — Trims whitespace around the figure. Almost always desirable.
- `facecolor` — Background color of the saved image.
- `transparent=True` — Saves with a transparent background (useful for overlaying on slides).

```python
# Save as PDF for publication-quality output
x = np.linspace(-3, 3, 300)
gaussian = np.exp(-x ** 2)

plt.plot(x, gaussian, linewidth=2)
plt.xlabel('x')
plt.ylabel('exp(-x²)')
plt.title('Gaussian Curve')

plt.savefig('data/gaussian.pdf', bbox_inches='tight')
plt.show()

print('Figure saved to data/gaussian.pdf')
```

## Quick Format Strings

`plt.plot()` accepts a third argument — a format string that compactly specifies color, marker, and line style. This is borrowed from MATLAB syntax.

Format: `'[color][marker][linestyle]'`

Examples: `'r--'` (red dashed), `'bo'` (blue circles), `'g^-'` (green triangles with solid line)

```python
x = np.linspace(0, 5, 20)

plt.plot(x, x, 'r--', label='linear')        # red dashed
plt.plot(x, x ** 1.5, 'bs', label='x^1.5')   # blue squares
plt.plot(x, x ** 2, 'g^', label='quadratic')  # green triangles

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Format String Examples')
plt.legend()
plt.show()
```

## Practical Example: Comparing Functions

Bringing together labels, legends, limits, and formatting to build a complete visualization.

```python
x = np.linspace(0, 3, 200)

plt.plot(x, x, 'b-', label='Linear: f(x) = x')
plt.plot(x, x ** 2, 'r--', label='Quadratic: f(x) = x²')
plt.plot(x, np.sqrt(x), 'g-.', label='Square root: f(x) = √x')
plt.plot(x, np.log1p(x), 'm:', label='Log: f(x) = ln(1+x)')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Comparing Growth Rates')
plt.xlim(0, 3)
plt.ylim(0, 5)
plt.legend(loc='upper left')
plt.show()
```

## Summary

| Function | Purpose |
|----------|---------|
| `plt.plot(x, y)` | Draw a line plot |
| `plt.xlabel()` / `plt.ylabel()` | Set axis labels |
| `plt.title()` | Set the plot title |
| `plt.xlim()` / `plt.ylim()` | Set axis ranges |
| `plt.legend()` | Display a legend |
| `plt.savefig()` | Export figure to file |
| `plt.show()` | Render and finalize the figure |

This functional (`pyplot`) interface is convenient for quick exploratory plots, but for more control — especially when working with multiple subplots — the **object-oriented** interface covered in the next notebook is preferred.
