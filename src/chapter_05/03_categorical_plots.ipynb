# Categorical Plots with Seaborn

Many real-world analyses involve categorical variables — days of the week, product categories, experimental groups. Seaborn offers a rich set of functions for visualizing how a numeric variable varies across categories.

This notebook covers:
- `sns.countplot()` — frequency counts per category
- `sns.barplot()` — means with confidence intervals
- `sns.boxplot()` — five-number summaries and outliers
- `sns.violinplot()` — full distribution shapes
- `sns.swarmplot()` and `sns.stripplot()` — individual observations
- `sns.catplot()` — figure-level categorical plotting
- Combining multiple plot types on the same axes

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
sns.set_theme(style='whitegrid')
```

```python
tips = sns.load_dataset('tips')
titanic = sns.load_dataset('titanic').dropna(subset=['age'])

print(f'tips: {tips.shape}')
print(f'titanic: {titanic.shape}')
tips.head()
```

## Count Plots: `sns.countplot()`

`countplot()` is essentially a histogram for categorical data — it shows the number of observations in each category.

```python
plt.figure(figsize=(7, 4))
sns.countplot(data=tips, x='day', order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Number of Meals per Day')
plt.show()
```

```python
# Count plot with hue for sub-groups
plt.figure(figsize=(8, 5))
sns.countplot(data=tips, x='day', hue='sex', order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Meals per Day by Sex')
plt.show()
```

```python
# Horizontal count plot — useful when category labels are long
plt.figure(figsize=(7, 4))
sns.countplot(data=titanic, y='class', hue='survived', palette='Set2')
plt.title('Titanic Passengers by Class and Survival')
plt.show()
```

## Bar Plots: `sns.barplot()`

`barplot()` shows the **mean** (or another estimator) of a numeric variable for each category, with error bars representing a 95% confidence interval by default.

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=tips, x='day', y='total_bill', order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Average Total Bill by Day (with 95% CI)')
plt.show()
```

```python
# Bar plot with hue and a custom estimator
plt.figure(figsize=(8, 5))
sns.barplot(
    data=tips, x='day', y='tip',
    hue='sex', estimator=np.median,
    order=['Thur', 'Fri', 'Sat', 'Sun'],
    palette='coolwarm',
)
plt.title('Median Tip by Day and Sex')
plt.show()
```

## Box Plots: `sns.boxplot()`

Box plots display the **median**, **interquartile range** (IQR, the box), **whiskers** (1.5x IQR), and **outliers** beyond the whiskers. They are an efficient way to compare spread and skew across groups.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=tips, x='day', y='total_bill', order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Total Bill Distribution by Day')
plt.show()
```

```python
# Box plot with hue
plt.figure(figsize=(9, 5))
sns.boxplot(
    data=tips, x='day', y='total_bill',
    hue='smoker', order=['Thur', 'Fri', 'Sat', 'Sun'],
    palette='pastel',
)
plt.title('Total Bill by Day and Smoker Status')
plt.show()
```

```python
# Horizontal box plot for Titanic ages by class
plt.figure(figsize=(9, 5))
sns.boxplot(data=titanic, x='age', y='class', palette='Set3')
plt.title('Passenger Age Distribution by Class')
plt.show()
```

## Violin Plots: `sns.violinplot()`

A violin plot combines a box plot with a mirrored KDE, showing the full distribution shape. It reveals multimodality and skew that box plots may hide.

```python
plt.figure(figsize=(9, 5))
sns.violinplot(data=tips, x='day', y='total_bill', order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Total Bill Violin Plot by Day')
plt.show()
```

```python
# Split violin to compare two groups side by side within each category
plt.figure(figsize=(9, 5))
sns.violinplot(
    data=tips, x='day', y='total_bill',
    hue='sex', split=True,
    order=['Thur', 'Fri', 'Sat', 'Sun'],
    palette='muted',
    inner='quartile',
)
plt.title('Split Violin: Total Bill by Day and Sex')
plt.show()
```

## Individual Observations: `sns.stripplot()` and `sns.swarmplot()`

- `stripplot()` — jittered points for each observation.
- `swarmplot()` — points are positioned so they do not overlap, giving a better sense of the distribution. Slower for large datasets.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.stripplot(data=tips, x='day', y='total_bill', ax=axes[0], alpha=0.6, jitter=True)
axes[0].set_title('Strip Plot')

sns.swarmplot(data=tips, x='day', y='total_bill', ax=axes[1], size=4)
axes[1].set_title('Swarm Plot')

plt.tight_layout()
plt.show()
```

```python
# Swarm plot with hue
plt.figure(figsize=(9, 5))
sns.swarmplot(
    data=tips, x='day', y='tip',
    hue='sex', dodge=True,
    order=['Thur', 'Fri', 'Sat', 'Sun'],
    size=4, palette='Set1',
)
plt.title('Individual Tips by Day and Sex')
plt.show()
```

## Combining Plot Types

Because axes-level functions draw on an existing `Axes`, you can layer them. A common combination is a violin (or box) plot with a strip or swarm plot on top.

```python
# Violin + Swarm
plt.figure(figsize=(9, 5))
sns.violinplot(
    data=tips, x='day', y='total_bill',
    order=['Thur', 'Fri', 'Sat', 'Sun'],
    inner=None, palette='pastel',
)
sns.swarmplot(
    data=tips, x='day', y='total_bill',
    order=['Thur', 'Fri', 'Sat', 'Sun'],
    color='black', size=3, alpha=0.6,
)
plt.title('Violin + Swarm: Total Bill by Day')
plt.show()
```

```python
# Box + Strip
plt.figure(figsize=(9, 5))
sns.boxplot(
    data=tips, x='day', y='tip',
    order=['Thur', 'Fri', 'Sat', 'Sun'],
    palette='light:#5A9', fliersize=0,  # hide box outlier markers
)
sns.stripplot(
    data=tips, x='day', y='tip',
    order=['Thur', 'Fri', 'Sat', 'Sun'],
    color='0.3', size=3, alpha=0.4, jitter=True,
)
plt.title('Box + Strip: Tip by Day')
plt.show()
```

## Figure-level Categorical: `sns.catplot()`

`catplot()` is the figure-level function that provides access to all categorical plot types via the `kind` parameter. It returns a `FacetGrid`, enabling `col` and `row` faceting.

Valid `kind` values: `'strip'`, `'swarm'`, `'box'`, `'violin'`, `'bar'`, `'count'`, `'point'`.

```python
# Faceted box plot
g = sns.catplot(
    data=tips, x='day', y='total_bill',
    col='time', kind='box',
    order=['Thur', 'Fri', 'Sat', 'Sun'],
    height=4, aspect=1.2,
)
g.fig.suptitle('Total Bill by Day — Faceted by Meal Time', y=1.03)
plt.show()
```

```python
# Faceted violin plot with hue
g = sns.catplot(
    data=tips, x='sex', y='tip',
    hue='smoker', col='day',
    kind='violin', split=True,
    col_wrap=2, height=4, aspect=1.1,
    palette='Set2',
)
g.fig.suptitle('Tip Distributions by Sex, Smoker Status, and Day', y=1.03)
plt.show()
```

```python
# Faceted count plot on Titanic data
g = sns.catplot(
    data=titanic, x='class', hue='survived',
    col='sex', kind='count',
    height=4, aspect=1,
    palette='coolwarm',
)
g.fig.suptitle('Titanic Survival Counts by Class and Sex', y=1.03)
plt.show()
```

## Key Takeaways

| Function | Shows | Best For |
|---|---|---|
| `countplot()` | Frequency counts | Seeing how many observations per category |
| `barplot()` | Mean + CI | Comparing central tendency across groups |
| `boxplot()` | Median, IQR, outliers | Quick summary of spread and skew |
| `violinplot()` | Full distribution shape | Revealing multimodality |
| `stripplot()` | Jittered points | Showing every observation (small-medium data) |
| `swarmplot()` | Non-overlapping points | Same as strip but avoids overlap |
| `catplot()` | Any of the above | Faceted categorical comparisons |

- Combine violin or box plots with strip/swarm overlays for both summary and detail.
- Use `split=True` in `violinplot()` to compare two hue levels side by side.
- `catplot()` is the preferred entry point when you need faceting.
