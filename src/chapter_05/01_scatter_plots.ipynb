# Scatter Plots with Seaborn

Scatter plots are one of the most fundamental tools for exploring relationships between two continuous variables. Seaborn provides several functions for creating scatter plots, each with increasing levels of sophistication — from simple point clouds to regression-fitted plots with marginal distributions.

In this notebook we cover:
- Setting visual themes with `sns.set_theme()` and `sns.set_style()`
- `sns.scatterplot()` with `hue`, `size`, and `style` encodings
- `sns.relplot()` for faceted scatter plots
- `sns.lmplot()` for scatter plots with regression lines
- `sns.jointplot()` for scatter plots with marginal distributions
- `sns.pairplot()` for visualizing all pairwise relationships

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
sns.set_theme()
```

## Loading Built-in Datasets

Seaborn ships with several practice datasets accessible via `sns.load_dataset()`. We will use **tips** (restaurant tipping data) and **penguins** (Palmer Archipelago penguin measurements) throughout this notebook.

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
penguins = sns.load_dataset('penguins').dropna()
penguins.head()
```

## Setting Visual Themes

Seaborn offers `set_theme()` to apply a consistent look and `set_style()` to change just the axes style. The five built-in styles are: `darkgrid`, `whitegrid`, `dark`, `white`, and `ticks`.

```python
# Compare styles side by side
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

styles = ['whitegrid', 'darkgrid', 'ticks']
for ax, style in zip(axes, styles):
    with sns.axes_style(style):
        ax.plot([0, 1, 2, 3], [0, 1, 4, 9])
        ax.set_title(f'Style: {style}')

plt.tight_layout()
plt.show()
```

```python
# Reset to the default theme for the rest of the notebook
sns.set_theme(style='whitegrid')
```

## Basic Scatter Plot with `sns.scatterplot()`

`scatterplot()` is the axes-level function for creating scatter plots. It draws points on a pair of continuous axes and supports mapping additional variables to visual properties like color, size, and marker shape.

```python
# Simple scatter plot: total bill vs tip
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Total Bill vs Tip Amount')
plt.show()
```

### Using `hue` to Encode a Third Variable

The `hue` parameter maps a categorical (or continuous) column to point color, letting you see group differences at a glance.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker')
plt.title('Total Bill vs Tip — Colored by Smoker Status')
plt.show()
```

### Using `size` and `style`

- `size` maps a variable to the area of each point — useful for a third continuous dimension.
- `style` maps a variable to marker shape — best for a small number of categories.

```python
plt.figure(figsize=(9, 6))
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='day',
    size='size',
    style='sex',
    sizes=(30, 200),
    alpha=0.7,
)
plt.title('Multi-dimensional Scatter: Bill, Tip, Day, Party Size, Sex')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Scatter Plot with Penguins

The penguins dataset has clear clusters by species, making it ideal for demonstrating `hue`.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=penguins,
    x='flipper_length_mm',
    y='body_mass_g',
    hue='species',
    style='species',
)
plt.title('Penguin Flipper Length vs Body Mass')
plt.show()
```

## Figure-level Scatter with `sns.relplot()`

`relplot()` is the **figure-level** counterpart to `scatterplot()`. Its key advantage is the `col` and `row` parameters, which create faceted grids of scatter plots. By default, `relplot()` draws scatter plots (`kind='scatter'`).

```python
# Faceted scatter plot: one column per time of day
g = sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',
    col='time',
    kind='scatter',
    height=4,
    aspect=1.2,
)
g.fig.suptitle('Tips by Time of Day', y=1.03)
plt.show()
```

```python
# Using both col and row for a two-way facet
g = sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',
    col='day',
    row='time',
    kind='scatter',
    height=3.5,
    aspect=1,
)
g.fig.suptitle('Tips Faceted by Day and Meal Time', y=1.02)
plt.show()
```

## Scatter with Regression: `sns.lmplot()`

`lmplot()` overlays a linear regression line (with a confidence interval) on top of a scatter plot. It is figure-level and supports `hue`, `col`, and `row` for multi-group regression comparisons.

```python
# Basic regression scatter
sns.lmplot(
    data=tips,
    x='total_bill',
    y='tip',
    height=5,
    aspect=1.3,
)
plt.title('Linear Fit: Total Bill vs Tip')
plt.show()
```

```python
# Separate regression lines per group
sns.lmplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',
    height=5,
    aspect=1.3,
)
plt.title('Regression Lines by Smoker Status')
plt.show()
```

```python
# Faceted regression by day
sns.lmplot(
    data=tips,
    x='total_bill',
    y='tip',
    col='day',
    hue='sex',
    col_wrap=2,
    height=4,
    aspect=1.1,
)
plt.show()
```

### Polynomial Regression with `order`

Set `order` to fit a higher-degree polynomial instead of a straight line.

```python
sns.lmplot(
    data=tips,
    x='total_bill',
    y='tip',
    order=2,
    height=5,
    aspect=1.3,
)
plt.title('Quadratic Fit: Total Bill vs Tip')
plt.show()
```

## Joint Distributions: `sns.jointplot()`

`jointplot()` combines a central scatter (or other bivariate plot) with marginal histograms or KDE curves on the axes. This lets you see the individual distributions alongside the relationship.

```python
# Default: scatter with marginal histograms
sns.jointplot(data=tips, x='total_bill', y='tip', height=6)
plt.show()
```

```python
# KDE contour joint plot
sns.jointplot(
    data=tips,
    x='total_bill',
    y='tip',
    kind='kde',
    fill=True,
    height=6,
)
plt.show()
```

```python
# Hex-bin joint plot — useful for large datasets to avoid overplotting
sns.jointplot(
    data=tips,
    x='total_bill',
    y='tip',
    kind='hex',
    height=6,
    color='teal',
)
plt.show()
```

```python
# Regression joint plot with hue
sns.jointplot(
    data=penguins,
    x='flipper_length_mm',
    y='body_mass_g',
    hue='species',
    height=7,
)
plt.show()
```

## Pairwise Relationships: `sns.pairplot()`

`pairplot()` creates a grid of scatter plots for every pair of numeric columns in a DataFrame, with histograms (or KDEs) on the diagonal. It is the quickest way to get an overview of multivariate structure.

```python
# Basic pairplot of the penguins dataset
sns.pairplot(penguins, hue='species', height=2.5)
plt.suptitle('Penguin Measurements — Pairwise Scatter', y=1.02)
plt.show()
```

```python
# Pairplot with KDE on the diagonal and selected columns
sns.pairplot(
    penguins,
    vars=['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm'],
    hue='species',
    diag_kind='kde',
    plot_kws={'alpha': 0.6},
    height=3,
)
plt.show()
```

```python
# Pairplot on tips dataset — a different look
sns.pairplot(
    tips,
    hue='time',
    diag_kind='hist',
    corner=True,  # show only lower triangle
    height=2.5,
)
plt.show()
```

## Key Takeaways

| Function | Level | Best For |
|---|---|---|
| `scatterplot()` | Axes | Single scatter plot on an existing axes |
| `relplot(kind='scatter')` | Figure | Faceted scatter grids via `col`/`row` |
| `lmplot()` | Figure | Scatter + regression with optional faceting |
| `jointplot()` | Figure | Bivariate scatter with marginal distributions |
| `pairplot()` | Figure | Quick overview of all pairwise relationships |

- **Axes-level** functions (`scatterplot`) draw onto a specific `matplotlib.axes.Axes` and return it.
- **Figure-level** functions (`relplot`, `lmplot`, `jointplot`, `pairplot`) create their own `Figure` and return a `FacetGrid` (or `JointGrid`).
- Use `hue`, `size`, and `style` to encode additional dimensions without creating separate plots.
