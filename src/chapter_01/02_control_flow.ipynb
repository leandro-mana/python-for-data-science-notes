## Conditional Statements: `if` / `elif` / `else`

Python uses indentation (4 spaces by convention) to define code blocks.
There are no braces or `end` keywords.

````python
temperature = 22

if temperature > 30:
    print('Hot day — stay hydrated!')
elif temperature > 20:
    print('Pleasant weather.')
elif temperature > 10:
    print('A bit cool — grab a jacket.')
else:
    print('Cold — bundle up!')
````

````python
# Combining conditions with logical operators
age = 25
has_license = True

if age >= 18 and has_license:
    print('You can drive.')
elif age >= 18 and not has_license:
    print('You are old enough, but you need a license.')
else:
    print('You are not old enough to drive.')
````

> **Pro tip:** Python evaluates `and` / `or` with short-circuit logic.
> In `A and B`, if `A` is falsy Python never evaluates `B`.
> In `A or B`, if `A` is truthy Python never evaluates `B`.

## `for` Loops

A `for` loop iterates over any **iterable** — lists, strings, tuples, dicts, sets,
ranges, files, and more.

````python
# Iterating over a list
languages = ['Python', 'R', 'SQL', 'Julia']

for lang in languages:
    print(f'{lang} has {len(lang)} characters')
````

````python
# Iterating over a string
for char in 'Data':
    print(char, end=' ')
print()  # newline

# Iterating over a dictionary
scores = {'Alice': 92, 'Bob': 85, 'Carol': 97}

for name, score in scores.items():
    status = 'Pass' if score >= 90 else 'Review'
    print(f'{name}: {score} ({status})')
````

## `while` Loops

A `while` loop repeats as long as its condition is truthy. Be careful to ensure
the condition eventually becomes false, or you will create an infinite loop.

````python
countdown = 5

while countdown > 0:
    print(countdown, end=' ')
    countdown -= 1

print('\nLiftoff!')
````

````python
# Practical example: simple convergence check
value = 100.0
tolerance = 0.01
iterations = 0

while value > tolerance:
    value *= 0.5
    iterations += 1

print(f'Converged after {iterations} iterations (value = {value:.6f})')
````

## `break`, `continue`, and `pass`

- **`break`** — exit the loop immediately
- **`continue`** — skip the rest of the current iteration and proceed to the next
- **`pass`** — do nothing (placeholder for empty code blocks)

````python
# break: stop searching once we find a target
data = [4, 7, 2, 9, 1, 5, 8]

for value in data:
    if value == 9:
        print(f'Found target: {value}')
        break
    print(f'  Checked {value}...')
````

````python
# continue: skip odd numbers
for i in range(10):
    if i % 2 != 0:
        continue
    print(i, end=' ')
print()
````

## `range()`

`range(start, stop, step)` generates an immutable sequence of integers.
It is memory-efficient because it computes values on the fly rather than
storing them in a list.

````python
# range(stop)
print(list(range(5)))          # [0, 1, 2, 3, 4]

# range(start, stop)
print(list(range(2, 8)))       # [2, 3, 4, 5, 6, 7]

# range(start, stop, step)
print(list(range(0, 20, 3)))   # [0, 3, 6, 9, 12, 15, 18]

# Counting backwards
print(list(range(10, 0, -2)))  # [10, 8, 6, 4, 2]
````

## `enumerate()`

When you need both the index and the value during iteration, use `enumerate()`
instead of manually managing a counter variable.

````python
tools = ['pandas', 'numpy', 'matplotlib', 'scikit-learn']

for index, tool in enumerate(tools):
    print(f'{index}: {tool}')

print()

# Start counting from 1
for position, tool in enumerate(tools, start=1):
    print(f'{position}. {tool}')
````

## `zip()`

`zip()` pairs up elements from two or more iterables. It stops at the shortest
iterable (use `itertools.zip_longest` if you need to handle unequal lengths).

````python
names = ['Alice', 'Bob', 'Carol']
scores = [92, 85, 97]
grades = ['A', 'B', 'A+']

for name, score, grade in zip(names, scores, grades):
    print(f'{name}: {score} ({grade})')

# Creating a dictionary from two lists
name_to_score = dict(zip(names, scores))
print('\nDictionary:', name_to_score)
````

## List Comprehensions

List comprehensions provide a concise way to create lists. The general syntax is:

```python
[expression for item in iterable if condition]
```

They are typically faster than equivalent `for` loop + `append()` patterns.

````python
# Squares of numbers 0 through 9
squares = [x ** 2 for x in range(10)]
print('Squares:', squares)

# With a filter: only even squares
even_squares = [x ** 2 for x in range(10) if x % 2 == 0]
print('Even squares:', even_squares)

# Transform strings
words = ['hello', 'world', 'python']
upper_words = [w.upper() for w in words]
print('Uppercased:', upper_words)
````

## Dictionary and Set Comprehensions

The same comprehension syntax works for dictionaries (using `{key: value ...}`)
and sets (using `{value ...}`).

````python
# Dict comprehension: word -> length
words = ['Python', 'is', 'great', 'for', 'data', 'science']
word_lengths = {word: len(word) for word in words}
print('Word lengths:', word_lengths)

# Dict comprehension with filter
long_words = {word: length for word, length in word_lengths.items() if length > 3}
print('Long words:  ', long_words)

# Set comprehension: unique first letters
first_letters = {word[0].lower() for word in words}
print('First letters:', first_letters)
````

## Nested Comprehensions

You can nest loops inside a comprehension. The order of the `for` clauses
matches the order you would write them in a traditional nested loop.

````python
# Flatten a matrix
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [element for row in matrix for element in row]
print('Flattened:', flat)

# Create a matrix with a nested comprehension
multiplication_table = [[i * j for j in range(1, 6)] for i in range(1, 6)]
print('\nMultiplication table (5x5):')
for row in multiplication_table:
    print(row)
````

````python
# Practical example: all (row, col) pairs where the value is even
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

even_positions = [
    (r, c)
    for r, row in enumerate(matrix)
    for c, val in enumerate(row)
    if val % 2 == 0
]
print('Positions of even values:', even_positions)
````

> **Pro tip:** If a comprehension becomes hard to read (more than ~80 characters
> or multiple nested loops with conditions), refactor it into a regular `for` loop.
> Readability counts.

## Ternary (Conditional) Expressions

Python's ternary expression lets you write a simple `if/else` on a single line:

```python
value_if_true if condition else value_if_false
```

````python
score = 85
result = 'Pass' if score >= 60 else 'Fail'
print(f'Score {score}: {result}')

# Ternary inside a comprehension
numbers = [1, -2, 3, -4, 5, -6]
labels = ['positive' if n > 0 else 'negative' for n in numbers]
print(labels)

# Ternary for concise variable assignment
x = None
value = x if x is not None else 'default'
print(f'value = {value!r}')
````
