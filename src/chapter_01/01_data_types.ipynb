## Numbers

Python has three built-in numeric types: **integers** (`int`), **floating-point numbers** (`float`),
and **complex numbers** (`complex`). Integers have arbitrary precision — there is no overflow.

```python
# Integers
x = 42
print(type(x), x)

# Floats
y = 3.14
print(type(y), y)

# Complex numbers (j is the imaginary unit)
z = 2 + 3j
print(type(z), z, '| real part:', z.real, '| imaginary part:', z.imag)
```

```python
# Arithmetic operators
a, b = 17, 5

print('Addition:      ', a + b)
print('Subtraction:   ', a - b)
print('Multiplication:', a * b)
print('True division: ', a / b)   # always returns float
print('Floor division:', a // b)  # truncates toward negative infinity
print('Modulo:        ', a % b)
print('Exponent:      ', a ** b)
```

> **Pro tip:** Use `divmod(a, b)` when you need both the quotient and remainder at once —
> it is slightly faster than computing `a // b` and `a % b` separately.

```python
quotient, remainder = divmod(17, 5)
print(f'17 divided by 5 is {quotient} remainder {remainder}')
```

## Strings

Strings are **immutable** sequences of Unicode characters. You can create them with single quotes,
double quotes, or triple quotes for multi-line text.

```python
greeting = 'Hello'
name = "World"
paragraph = """This is a
multi-line string."""

print(greeting)
print(name)
print(paragraph)
```

### String Indexing and Slicing

Strings support zero-based indexing and slicing with the syntax `s[start:stop:step]`.

```python
s = 'Data Science'

print('First character: ', s[0])
print('Last character:  ', s[-1])
print('First 4 chars:  ', s[:4])
print('From index 5:   ', s[5:])
print('Every 2nd char: ', s[::2])
print('Reversed:       ', s[::-1])
```

### Useful String Methods

```python
text = '  python for Data Science  '

print('strip():     ', repr(text.strip()))
print('lower():     ', text.strip().lower())
print('upper():     ', text.strip().upper())
print('title():     ', text.strip().title())
print('replace():   ', text.strip().replace('Data', 'Applied'))
print('split():     ', text.split())
print('find("Data"):', text.find('Data'))
print('count("a"):  ', text.count('a'))
print('startswith():', text.strip().startswith('python'))
print('isdigit():   ', '42'.isdigit())
```

### f-Strings (Formatted String Literals)

Introduced in Python 3.6, f-strings are the recommended way to embed expressions inside strings.

```python
language = 'Python'
version = 3.12

# Basic embedding
print(f'{language} {version}')

# Expressions inside braces
print(f'Sum: {10 + 20}')

# Format specifiers
pi = 3.141592653589793
print(f'Pi to 2 decimals: {pi:.2f}')
print(f'Large number:     {1_000_000:,}')
print(f'Percentage:       {0.856:.1%}')
print(f'Left-padded:      {42:>10}')
```

## Booleans

Booleans represent truth values: `True` and `False`. They are a subclass of `int`
(where `True == 1` and `False == 0`). Python evaluates "truthy" and "falsy" values in
boolean contexts — empty containers and zero are falsy, everything else is truthy.

```python
print(True + True)          # 2, because True is 1
print(isinstance(True, int)) # True

# Falsy values
falsy_values = [0, 0.0, '', [], {}, set(), None, False]
print('All falsy?', all(not v for v in falsy_values))

# Logical operators
print(True and False)   # False
print(True or False)    # True
print(not True)         # False
```

## Lists

Lists are **mutable**, ordered sequences. They can hold items of any type and are one of
the most versatile data structures in Python.

```python
# Creating lists
numbers = [10, 20, 30, 40, 50]
mixed = [1, 'two', 3.0, True, None]

print('numbers:', numbers)
print('mixed:  ', mixed)
print('length: ', len(numbers))
```

```python
# Indexing and slicing (same syntax as strings)
print('First:   ', numbers[0])
print('Last:    ', numbers[-1])
print('Slice:   ', numbers[1:4])
print('Reversed:', numbers[::-1])
```

### List Methods

```python
fruits = ['apple', 'banana', 'cherry']

# Adding elements
fruits.append('date')            # add to end
fruits.insert(1, 'avocado')      # insert at index
fruits.extend(['elderberry'])     # extend with iterable
print('After adding:', fruits)

# Removing elements
fruits.remove('banana')          # remove first occurrence
popped = fruits.pop()            # remove and return last item
print('Popped:', popped)
print('After removing:', fruits)

# Sorting and reversing
nums = [3, 1, 4, 1, 5, 9, 2, 6]
nums.sort()
print('Sorted:', nums)
nums.reverse()
print('Reversed:', nums)
```

### Nested Lists

```python
# A matrix represented as a list of lists
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print('Row 0:      ', matrix[0])
print('Element 1,2:', matrix[1][2])  # row 1, column 2 -> 6

# Iterate over rows
for row in matrix:
    print(row)
```

## Tuples

Tuples are **immutable** ordered sequences. Use them when you need a fixed collection
of items — for example, returning multiple values from a function, or as dictionary keys.

```python
point = (3, 4)
single = (42,)   # trailing comma is required for a single-element tuple

print('point:', point, '| type:', type(point))
print('single:', single, '| type:', type(single))

# Unpacking
x, y = point
print(f'x={x}, y={y}')

# Tuples are immutable
try:
    point[0] = 10
except TypeError as e:
    print(f'Error: {e}')

# Useful methods
t = (1, 2, 2, 3, 3, 3)
print('count(3):', t.count(3))
print('index(2):', t.index(2))
```

## Sets

Sets are **unordered** collections of **unique** elements. They support fast membership
testing and mathematical set operations (union, intersection, difference).

```python
# Creating sets
colors = {'red', 'green', 'blue'}
from_list = set([1, 2, 2, 3, 3, 3])

print('colors:   ', colors)
print('from_list:', from_list)  # duplicates removed

# Adding and removing
colors.add('yellow')
colors.discard('red')  # discard does not raise error if missing
print('Updated:', colors)
```

```python
# Set operations
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print('Union:        ', a | b)        # or a.union(b)
print('Intersection: ', a & b)        # or a.intersection(b)
print('Difference:   ', a - b)        # or a.difference(b)
print('Sym Difference:', a ^ b)       # or a.symmetric_difference(b)
print('Is subset:    ', {1, 2}.issubset(a))
```

## Dictionaries

Dictionaries are **mutable** mappings of key-value pairs. Since Python 3.7, they
preserve insertion order. Keys must be hashable (typically strings, numbers, or tuples).

```python
student = {
    'name': 'Alice',
    'age': 28,
    'courses': ['Statistics', 'Machine Learning']
}

print('Name: ', student['name'])
print('Age:  ', student.get('age'))
print('Grade:', student.get('grade', 'N/A'))  # default if missing
```

### Dictionary Methods

```python
# Adding and updating
student['grade'] = 'A'
student.update({'age': 29, 'university': 'MIT'})
print('Updated:', student)

# Removing
removed = student.pop('university')
print(f'Removed "{removed}":', student)

# Iterating
print('\nKeys and values:')
for key, value in student.items():
    print(f'  {key}: {value}')
```

### Nested Dictionaries

```python
school = {
    'student_001': {
        'name': 'Alice',
        'scores': {'math': 92, 'english': 88}
    },
    'student_002': {
        'name': 'Bob',
        'scores': {'math': 75, 'english': 95}
    }
}

# Accessing nested values
print(school['student_001']['scores']['math'])  # 92

# Iterate and summarize
for sid, info in school.items():
    avg = sum(info['scores'].values()) / len(info['scores'])
    print(f"{info['name']}: average score = {avg:.1f}")
```

## Type Conversion

Python provides built-in functions to convert between types. This is especially
common when reading data from files or user input (which arrives as strings).

```python
# Numeric conversions
print(int(3.9))        # 3  (truncates, does NOT round)
print(float(7))        # 7.0
print(complex(2, 5))   # (2+5j)

# String <-> Number
print(int('42'))       # 42
print(float('3.14'))   # 3.14
print(str(100))        # '100'

# Collection conversions
print(list('hello'))            # ['h', 'e', 'l', 'l', 'o']
print(tuple([1, 2, 3]))        # (1, 2, 3)
print(set([1, 1, 2, 3]))       # {1, 2, 3}
print(list({1: 'a', 2: 'b'}))  # [1, 2]  (keys)

# Boolean conversion
print(bool(0), bool(''), bool([]))       # False False False
print(bool(1), bool('hi'), bool([0]))    # True True True
```

## Comparison Operators

Comparison operators return booleans. Python supports chained comparisons,
which makes range checks very readable.

```python
x = 10

print('x == 10:', x == 10)   # equal
print('x != 5: ', x != 5)    # not equal
print('x > 5:  ', x > 5)     # greater than
print('x <= 10:', x <= 10)   # less than or equal

# Chained comparison
print('1 < x < 20:', 1 < x < 20)

# Identity vs equality
a = [1, 2, 3]
b = [1, 2, 3]
print('a == b:', a == b)   # True  (same value)
print('a is b:', a is b)   # False (different objects)

# Membership
print('2 in a:   ', 2 in a)
print('4 not in a:', 4 not in a)
```
