## list comprehension

```python
x = [1, 2, 3, 4]
```

```python
# usual way without listcomprehension
out = []
for item in x:
    out.append(item**2)

print(out)
```

```python
# With listcomprehension
[item**2 for item in x]
```

```python
[item**2 for item in x if item % 2 == 0]
```

## lambda expressions

```python
def times2(var):
    return var * 2

times2(2)
```

```python
lambda var: var * 2
```

```python
result = lambda var: var * 2
result(4)
```

```python
print(f'two times two is: {result(2)}\nsix times two is: {result(6)}')
```

## map and filter

```python
seq = [1, 2, 3, 4, 5]
```

```python
# Apply the times2 function defined above to the whole sequence of numbers
map(times2, seq)
```

```python
# As above returns a Map type, by listing it returns the values
list(map(times2,seq))
```

```python
list(map(lambda var: var*2, seq))
```

```python
list(map(result, seq))
```

```python
# Now filtering ONLY the ones that are divisible by 2
filter(lambda item: item % 2 == 0, seq)
```

```python
list(filter(lambda item: item % 2 == 0, seq))
```
